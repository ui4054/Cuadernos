# 1. Magnitudes Escalares

Las magnitudes escalares son aquellas que quedan completamente definidas con un número y su correspondiente unidad de medida. Algunos ejemplos comunes de magnitudes escalares son la masa, la longitud, la temperatura y el tiempo.

## 1.1 Masa
La **masa** es una magnitud escalar que mide la cantidad de materia que posee un cuerpo. Su unidad en el Sistema Internacional (SI) es el kilogramo (kg).

## 1.2 Longitud
La **longitud** es una magnitud escalar que mide la distancia entre dos puntos. Su unidad en el SI es el metro (m).

## 1.3 Temperatura
La **temperatura** es una magnitud escalar que mide el grado de calor o frío de un cuerpo o del ambiente. Su unidad en el SI es el kelvin (K), aunque es común usar grados Celsius (°C).

## 1.4 Tiempo
El **tiempo** es una magnitud escalar que mide la duración de los eventos. Su unidad en el SI es el segundo (s).

# 4. Vectores

Un vector es una cantidad física que tiene tanto magnitud como dirección. Los vectores se representan gráficamente como flechas en un plano, donde la longitud de la flecha indica la magnitud y la dirección de la flecha indica la dirección del vector.

## 4.1 Representación de un Vector en el Plano

Para representar un vector en el plano, se utilizan sus componentes en las direcciones **x** e **y**. Si un vector tiene una magnitud **A** y forma un ángulo **θ** con el eje **x**, sus componentes son:

$$\{
A_x = A \cdot \cos(\theta)
\}
\{
A_y = A \cdot \sin(\theta)
\}$$

## 4.2 Teorema de Pitágoras Aplicado al Vector

El Teorema de Pitágoras se utiliza para calcular la magnitud de un vector a partir de sus componentes:

$$\{
|A| = \sqrt{A_x^2 + A_y^2}
\}$$

## 4.3 Suma y Resta de Vectores en Coordenadas Cartesianas

La suma de dos vectores **A** y **B** se realiza sumando sus componentes correspondientes:

$$\{
R_x = A_x + B_x
\}
\{
R_y = A_y + B_y
\}$$

De manera similar, la resta de vectores se realiza restando sus componentes correspondientes:

$$\{
D_x = A_x - B_x
\}
\{
D_y = A_y - B_y
\}$$

## 4.5 Coordenadas Polares

En coordenadas polares, un vector se representa por su magnitud y el ángulo que forma con el eje x. Las conversiones entre coordenadas cartesianas y polares se realizan de la siguiente manera:

\[
r = \sqrt{x^2 + y^2}
\]
\[
\theta = \tan^{-1}\left(\frac{y}{x}\right)
\]

## 4.6 Suma y Resta de Vectores en Coordenadas Polares

Para sumar o restar vectores en coordenadas polares, primero convierte los vectores a coordenadas cartesianas, realiza la operación y luego convierte el resultado de nuevo a coordenadas polares.

### Ejemplos Resueltos: Notación Científica y Notación de Ingeniería

#### Ejemplo 1: Convertir 1500 cm a metros

1. **Identificar la relación de conversión:** Sabemos que 1 metro = 100 centímetros.

2. **Convertir a notación científica:**
   $$\{
   1500 \text{ cm} = \frac{1500 \text{ cm}}{100 \text{ cm/m}} = 15 \text{ m}
   \}$$
   En notación científica:
   $$\{
   15 \text{ m} = 1.5 \times 10^1 \text{ m}
   \}$$

3. **Convertir a notación de ingeniería:**
   Dado que la notación de ingeniería debe expresar el número en múltiplos de 10^3, no es necesario ningún ajuste adicional:
   $$\{
   15 \text{ m} = 15 \times 10^0 \text{ m}
   \}$$

**Resultado:**
$$\{
1500 \text{ cm} = 15 \text{ m} = 1.5 
\}$$

---

#### Ejemplo 2: Convertir 0.0075 km a metros

1. **Identificar la relación de conversión:** Sabemos que 1 kilómetro = 1000 metros.

2. **Convertir a notación científica:**
   $$\{
   0.0075 \text{ km} = 0.0075 \times 1000 \text{ m/km} = 7.5 \text{ m}
   \}$$
   En notación científica:
   $$\{
   7.5 \text{ m} = 7.5 \times 10^0 \text{ m}
   \}$$

3. **Convertir a notación de ingeniería:**
   Nuevamente, no es necesario ajustar, ya que está en un múltiplo de 10^0:
   $$\{
   7.5 \text{ m} = 7.5 \times 10^0 \text{ m}
   \}$$

**Resultado:**
$$\{
0.0075 \text{ km} = 7.5 \text{ m} = 7.5 \times 10^0 \text{ m}
\}$$

### Ejemplos Resueltos: Suma y Resta de Vectores en Coordenadas Cartesianas

#### Ejemplo 1: Sumar **A** = (3, 4) y **B** = (1, 2)

1. **Sumar las componentes:** 
   $$\{
   R_x = A_x + B_x = 3 + 1 = 4
   \}$$
   $$\{
   R_y = A_y + B_y = 4 + 2 = 6
   \}$$

2. **Resultado final:** 
   $$\{
   \mathbf{R} = (4, 6)
   \}$$

**Resultado:**
$$\{
\mathbf{A} + \mathbf{B} = (4 + 6i)
\}$$

---

#### Ejemplo 2: Restar **A** = (5, -3) y **B** = (-2, 4)

1. **Restar las componentes:** 
   $$\{
   D_x = A_x - B_x = 5 - (-2) = 5 + 2 = 7
   \}$$
   $$\{
   D_y = A_y - B_y = -3 - 4 = -7
   \}$$

2. **Resultado final:** 
   $$\{
   \mathbf{D} = (7, -7)
   \}$$

**Resultado:**
$$\{
\mathbf{A} - \mathbf{B} = (7 - 7i)
\}$$

### Ejemplos Resueltos: Suma y Resta de Vectores en Coordenadas Polares

#### Ejemplo 1: Sumar **A** = (5, 30°) y **B** = (3, 45°)

1. **Convertir a coordenadas cartesianas:**
   $$\{
   A_x = 5 \cos(30°) = 5 \times 0.866 = 4.33
   \}$$
   $$\{
   A_y = 5 \sin(30°) = 5 \times 0.5 = 2.5
   \}$$
   $$\{
   B_x = 3 \cos(45°) = 3 \times 0.707 = 2.12
   \}$$
   $$\{
   B_y = 3 \sin(45°) = 3 \times 0.707 = 2.12
   \}$$

2. **Sumar las componentes:**
   $$\{
   R_x = 4.33 + 2.12 = 6.45
   \}$$
   $$\{
   R_y = 2.5 + 2.12 = 4.62
   \}$$

3. **Convertir de nuevo a coordenadas polares:**
   $$\{
   r = \sqrt{6.45^2 + 4.62^2} = \sqrt{41.60 + 21.35} = \sqrt{62.95} = 7.94
   \}$$
   $$\{
   \theta = \tan^{-1}\left(\frac{4.62}{6.45}\right) = \tan^{-1}(0.716) = 35.6°
   \}$$

**Resultado:**
$$\{
\mathbf{R} = (7.94, 35.6°)
\}$$

---

#### Ejemplo 2: Restar **A** = (7, 60°) y **B** = (2, 120°)

1. **Convertir a coordenadas cartesianas:**
   $$\{
   A_x = 7 \cos(60°) = 7 \times 0.5 = 3.5
   \}$$
   $$\{
   A_y = 7 \sin(60°) = 7 \times 0.866 = 6.06
   \}$$
   $$\{
   B_x = 2 \cos(120°) = 2 \times (-0.5) = -1
   \}$$
   $$\{
   B_y = 2 \sin(120°) = 2 \times 0.866 = 1.73
   \}$$

2. **Restar las componentes:**
   $$\{
   D_x = 3.5 - (-1) = 3.5 + 1 = 4.5
   \}$$
   $$\{
   D_y = 6.06 - 1.73 = 4.33
   \}$$

3. **Convertir de nuevo a coordenadas polares:**
   $$\{
   r = \sqrt{4.5^2 + 4.33^2} = \sqrt{20.25 + 18.75} = \sqrt{39} = 6.24
   \}$$
   $$\{
   \theta = \tan^{-1}\left(\frac{4.33}{4.5}\right) = \tan^{-1}(0.962) = 43.6°
   \}$$

**Resultado:**
$$\{
\mathbf{D} = (6.24, 43.6°)
\}$$
