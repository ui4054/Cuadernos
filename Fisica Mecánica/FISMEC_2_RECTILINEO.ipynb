# 1. Definición de Posición, Desplazamiento, Tiempo, Velocidad y Rapidez

## 1.1 Posición
La **posición** de un objeto es el lugar donde se encuentra en un determinado instante de tiempo respecto a un sistema de referencia. Se representa generalmente con el vector $$(\vec{r})$$.

**Ejemplo:**
- Si un coche está estacionado a 5 metros al este de un poste de luz, la posición del coche respecto al poste es de (5m) en la dirección este.

## 1.2 Desplazamiento
El **desplazamiento** es el cambio de posición de un objeto. Es una magnitud vectorial que indica la distancia y la dirección desde la posición inicial hasta la posición final.

**Ejemplo:**
- Si un ciclista se mueve desde un punto \(A\) hasta un punto \(B\), a 20 metros al norte de \(A\), el desplazamiento es de (20 m) hacia el norte.

## 1.3 Tiempo
El **tiempo** es una magnitud escalar que mide la duración de un evento o la separación entre dos instantes. Se mide en segundos (s) en el Sistema Internacional de Unidades (SI).

**Ejemplo:**
- Si un corredor tarda 10 segundos en completar una vuelta a la pista, el tiempo transcurrido es de (10s}).

## 1.4 Velocidad
La **velocidad** es una magnitud vectorial que describe la rapidez con la que un objeto cambia de posición, así como la dirección de ese cambio. La velocidad se define como el desplazamiento dividido por el tiempo.

$$[
\vec{v} = \frac{\vec{d}}{t}
]$$

**Ejemplo:**
- Si un automóvil se desplaza 100 metros hacia el norte en 20 segundos, su velocidad es de (5 m/s) hacia el norte.

## 1.5 Rapidez
La **rapidez** es una magnitud escalar que mide la distancia recorrida por un objeto en un cierto intervalo de tiempo, independientemente de la dirección.

$$[
\text{Rapidez} = \frac{\text{distancia recorrida}}{t}
]$$

## Ejemplo Simple:
- Si un corredor recorre 200 metros en 25 segundos, su rapidez es de (8m/s).

## Ejemplo en Python: Desplazamiento y Velocidad en un Plano Cartesiano

Supongamos que un objeto se mueve a lo largo de un plano cartesiano en varios intervalos de tiempo, cambiando su velocidad en cada intervalo. Vamos a calcular el desplazamiento total y la velocidad promedio del objeto.

### Datos:
- Intervalo 1: El objeto se mueve 3 metros hacia el este en 2 segundos.
- Intervalo 2: Luego, se mueve 4 metros hacia el norte en 3 segundos.
- Intervalo 3: Finalmente, se mueve 5 metros hacia el oeste en 4 segundos.

### Solución:

### 1. Desplazamiento en cada intervalo
Utilizaremos Python para graficar el movimiento del objeto en un plano cartesiano y calcular la velocidad en cada intervalo, así como la velocidad promedio.

```python
import matplotlib.pyplot as plt
import numpy as np

# Posiciones iniciales y tiempos
posiciones = [(0, 0), (3, 0), (3, 4), (-2, 4)]
tiempos = [2, 3, 4]

# Cálculo de desplazamientos y velocidades
desplazamientos = []
velocidades = []
distancias_acumuladas = [0]

for i in range(1, len(posiciones)):
    dx = posiciones[i][0] - posiciones[i-1][0]
    dy = posiciones[i][1] - posiciones[i-1][1]
    desplazamiento = np.sqrt(dx**2 + dy**2)
    velocidad = desplazamiento / tiempos[i-1]
    desplazamientos.append(desplazamiento)
    velocidades.append(velocidad)
    distancias_acumuladas.append(distancias_acumuladas[-1] + desplazamiento)

# Desplazamiento total
desplazamiento_total = np.sqrt((posiciones[-1][0] - posiciones[0][0])**2 + (posiciones[-1][1] - posiciones[0][1])**2)

# Velocidad promedio
tiempo_total = sum(tiempos)
velocidad_promedio = desplazamiento_total / tiempo_total # ecuación de velocidad promedio

# Graficar el movimiento en el plano cartesiano
x, y = zip(*posiciones)
plt.figure(figsize=(12, 5))

# Gráfica 1: Movimiento en el plano cartesiano
plt.subplot(1, 2, 1)
plt.plot(x, y, marker='o', color ='red')
plt.title("Movimiento del objeto en el plano cartesiano")
plt.xlabel("x (metros)")
plt.ylabel("y (metros)")
plt.grid(True)

# Gráfica 2: Distancia vs. Tiempo
tiempos_acumulados = np.cumsum([0] + tiempos)
plt.subplot(1, 2, 2)
plt.plot(tiempos_acumulados, distancias_acumuladas, marker='o')
plt.title("Gráfica de Distancia vs. Tiempo")
plt.xlabel("Tiempo (segundos)")
plt.ylabel("Distancia (metros)")
plt.grid(True)

plt.tight_layout()
plt.show()

# Resultados
print(f"Desplazamientos en cada intervalo: {desplazamientos}")
print(f"Velocidades en cada intervalo: {velocidades}")
print(f"Desplazamiento total: {desplazamiento_total:.2f} metros")
print(f"Velocidad promedio: {velocidad_promedio:.2f} m/s")
```

# Movimiento Rectilíneo Uniforme (M.R.U.)

## Definición
El **Movimiento Rectilíneo Uniforme (M.R.U.)** es un tipo de movimiento en el que un objeto se desplaza a lo largo de una línea recta con una **velocidad constante**. Esto significa que el objeto recorre distancias iguales en tiempos iguales, sin importar cuán corto o largo sea el intervalo de tiempo considerado.

### Características del M.R.U.
- **Trayectoria rectilínea**: El objeto se mueve en una línea recta.
- **Velocidad constante**: La velocidad no cambia con el tiempo. Es decir, no hay aceleración.
- **Distancia proporcional al tiempo**: La distancia recorrida es directamente proporcional al tiempo transcurrido.

## Ecuaciones Fundamentales
En el M.R.U., la relación entre la posición del objeto, el tiempo y la velocidad se describe por las siguientes ecuaciones:

1. **Ecuación de la posición**:
   $$[
   x(t) = x_0 + v \cdot t
   ]$$
   - (x(t)): Posición del objeto en el tiempo (t)
   - (x_0): Posición inicial del objeto (en (t = 0))
   - (v): Velocidad constante del objeto.
   - (t): Tiempo transcurrido.

2. **Ecuación de la velocidad**:
   $$\[
   v = \frac{d}{t}
   ]$$
   - (v): Velocidad constante del objeto.
   - (d): Distancia recorrida.
   - (t): Tiempo transcurrido.

## Gráficas del M.R.U.
En el M.R.U., las gráficas de posición vs. tiempo y velocidad vs. tiempo son líneas rectas:

1. **Gráfica de posición vs. tiempo**: 
   - Es una línea recta con pendiente positiva si la velocidad es positiva.
   - La pendiente de la recta representa la velocidad del objeto.

2. **Gráfica de velocidad vs. tiempo**: 
   - Es una línea recta horizontal, ya que la velocidad es constante.

## Ejemplos Comunes
- Un automóvil que viaja por una carretera recta a una velocidad constante.
- Un tren que se desplaza entre dos estaciones a velocidad constante.
- Un ciclista que mantiene una velocidad constante en una pista recta.

## Aplicaciones del M.R.U.
El M.R.U. es un modelo simplificado que se utiliza para describir el movimiento de objetos cuando la aceleración es despreciable o nula. Es útil para resolver problemas de cinemática y para comprender los conceptos básicos de movimiento en física.


# Ejemplos Resueltos de Movimiento Rectilíneo Uniforme (M.R.U.)

## Ejemplo 1: Cálculo de la Posición Final

### Enunciado
Un coche viaja a una velocidad constante de \( v = 20 \, \text{m/s} \) a lo largo de una carretera recta. Si la posición inicial del coche es \( x_0 = 50 \, \text{m} \), ¿cuál será su posición después de 10 segundos?

### Solución

Utilizamos la ecuación de la posición para el M.R.U.:

$$[
x(t) = x_0 + v \cdot t
]$$

Donde:
- ( x_0 = 50m )
- ( v = 20 {m/s})
- ( t = 10{segundos})

Sustituyendo los valores:

$$[
x(10) = 50 \, \text{m} + (20 \, \text{m/s}) \cdot (10 \, \text{s})
]$$

$$[
x(10) = 50 \, \text{m} + 200 \, \text{m} = 250 \, \text{m}
]$$

### Respuesta:
La posición final del coche después de 10 segundos es $$( x = 250 \, \text{m} )$$.


## Ejemplo 2: Cálculo de la Velocidad

### Enunciado
Un corredor recorre una distancia de \( 400 \, \text{m} \) en \( 50 \, \text{segundos} \) manteniendo una velocidad constante. ¿Cuál es la velocidad del corredor?

### Solución

Utilizamos la ecuación de la velocidad para el M.R.U.:

$$[
v = \frac{d}{t}
]$$

Donde:
-  d = 400 {m}
-  t = 50 {segundos}

Sustituyendo los valores:

$$[
v = \frac{400 \, \text{m}}{50 \, \text{s}} = 8 \, \text{m/s}
]$$

### Respuesta:
La velocidad del corredor es $$( v = 8 \, \text{m/s} )$$.

# Movimiento Circular Uniforme (M.C.U.)

El Movimiento Circular Uniforme (M.C.U.) es un tipo de movimiento en el cual un objeto se mueve a lo largo de una trayectoria circular con una velocidad constante. Aunque la velocidad en términos de magnitud es constante, la dirección del objeto cambia continuamente, lo que implica que hay una aceleración centrípeta hacia el centro del círculo.

## Conceptos Claves

1. **Velocidad Tangencial (\( v \)):**
   - La velocidad tangencial es la velocidad del objeto medida en la dirección tangente a la trayectoria circular en cualquier punto.
   - Se mantiene constante en el M.C.U.

2. **Frecuencia (\( f \)) y Periodo (\( T \)):**
   - **Frecuencia:** Número de revoluciones por unidad de tiempo. Se mide en revoluciones por segundo (Hz).
   - **Periodo:** Tiempo necesario para completar una revolución. Se mide en segundos (s).
   - La relación entre frecuencia y periodo es: **$$( f = \frac{1}{T} ) y ( T = \frac{1}{f} )$$**.

3. **Radio de la Trayectoria (\( r \)):**
   - El radio es la distancia constante desde el centro de la trayectoria circular hasta el objeto.

4. **Aceleración Centrífuga (\( a_c \)):**
   - La aceleración centrífuga es la aceleración que mantiene al objeto en su trayectoria circular. Se dirige hacia el centro del círculo.
   - Se calcula como: 
     $$[
     a_c = \frac{v^2}{r}
     ]$$
   donde \( v \) es la velocidad tangencial y \( r \) es el radio de la trayectoria.

5. **Velocidad Angular (w):**
   - La velocidad angular es la tasa a la cual el ángulo cambia con respecto al tiempo. Se mide en radianes por segundo (rad/s).
   - La relación entre velocidad angular y velocidad tangencial es:
     $$[
     v = \omega \cdot r
     ]$$
   - La velocidad angular se relaciona con el periodo mediante:
     $$[
     \omega = \frac{2\pi}{T}
     ]$$

# Soluciones a los Ejemplos del Movimiento Circular Uniforme (M.C.U.)

## Ejemplo 1 
### Un objeto se mueve a lo largo de una pista circular con un radio de 5 metros a una velocidad constante de 10 m/s. Calcular la aceleración centrífuga y la frecuencia del movimiento.

**Datos:**
- Velocidad tangencial (\(v\)) = 10 m/s
- Radio (\(r\)) = 5 m

**Cálculos:**

1. **Aceleración centrífuga (\(a_c\)):**
   $$
   a_c = \frac{v^2}{r}
   $$
   Sustituyendo los valores:
   $$
   a_c = \frac{(10 \, \text{m/s})^2}{5 \, \text{m}} = \frac{100 \, \text{m}^2/\text{s}^2}{5 \, \text{m}} = 20 \, \text{m/s}^2
   $$

2. **Frecuencia (\(f\)):**
   Primero, calculamos el periodo (\(T\)):
   $$
   T = \frac{2 \pi r}{v}
   $$
   Sustituyendo los valores:
   $$
   T = \frac{2 \pi \cdot 5 \, \text{m}}{10 \, \text{m/s}} = \frac{10 \pi}{10} = \pi \, \text{s}
   $$
   Ahora calculamos la frecuencia:
   $$
   f = \frac{1}{T} = \frac{1}{\pi} \approx 0.318 \, \text{Hz}
   $$

**Resultados:**
- **Aceleración centrífuga:** \(20 \,{m/s}^2\)
- **Frecuencia:** \(0.318 \, {Hz}\)

## Ejemplo 2 
### Un satélite orbita la Tierra a una velocidad tangencial de 7.5 km/s con un radio de 42,000 km. Determina la aceleración centrífuga y la velocidad angular del satélite.
**Datos:**
- Velocidad tangencial (\(v\)) = 7.5 km/s = 7500 m/s (conversión de km/s a m/s)
- Radio (\(r\)) = 42,000 km = 42,000,000 m (conversión de km a m)

**Cálculos:**

1. **Aceleración centrífuga (\(a_c\)):**
   $$
   a_c = \frac{v^2}{r}
   $$
   Sustituyendo los valores:
   $$
   a_c = \frac{(7500 \, \text{m/s})^2}{42,000,000 \, \text{m}} = \frac{56,250,000 \, \text{m}^2/\text{s}^2}{42,000,000 \, \text{m}} \approx 1.34 \, \text{m/s}^2
   $$

2. **Velocidad angular (\(\omega\)):**
   Primero, calculamos el periodo (\(T\)):
   $$
   T = \frac{2 \pi r}{v}
   $$
   Sustituyendo los valores:
   $$
   T = \frac{2 \pi \cdot 42,000,000 \, \text{m}}{7500 \, \text{m/s}} = \frac{264,000,000 \pi}{7500} \approx 110,000 \pi \approx 345,575 \, \text{s}
   $$
   Ahora calculamos la velocidad angular:
   $$
   \omega = \frac{2 \pi}{T} = \frac{2 \pi}{345,575} \approx 0.000018 \, \text{rad/s}
   $$

**Resultados:**
- **Aceleración centrífuga:** \(1.34 \, {m/s}^2\)
- **Velocidad angular:** \(0.000018 \, {rad/s}\)

### Ejercicio: Movimiento Circular Uniforme

**Enunciado:**

Considera un objeto que se mueve en un trayecto circular con un radio de \(1 \, \text{m}\). La velocidad tangencial del objeto es \(6.28 \, \text{m/s}\).

1. **Calcular el Periodo (\(T\))**: 
   El periodo es el tiempo que tarda el objeto en completar una vuelta completa alrededor del círculo. Utiliza la velocidad tangencial dada para calcularlo.

2. **Calcular la Velocidad Angular (\(\omega\))**: 
   La velocidad angular es la rapidez con la que el objeto gira alrededor del centro del círculo. Utiliza el periodo calculado para encontrarla.

3. **Graficar el Movimiento Circular y su Forma de Onda Senoidal**:
   - **Gráfica 1**: Movimiento circular del objeto en un plano cartesiano.
   - **Gráfica 2**: Proyección de la posición del objeto en el eje \(y\) en función del tiempo, mostrando la forma de onda senoidal resultante del movimiento circular.

**Instrucciones:**

1. Utiliza las siguientes fórmulas para los cálculos:
   - Periodo: $$(T = \frac{2 \pi r}{v})$$
   - Velocidad Angular: $$(\omega = \frac{2 \pi}{T})$$

2. Graficar el movimiento circular en el plano cartesiano usando la fórmula:
   - Componente en x $$(x(t) = r \cdot \cos(\omega t))$$
   - Componente en y $$(y(t) = r \cdot \sin(\omega t))$$

3. Graficar la proyección de la posición en el eje \(y\) como una función del tiempo, que representa una onda senoidal.

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos del movimiento circular
r = 1  # Radio en metros
v = 2*np.pi  # Velocidad tangencial en m/s
T = 2 * np.pi * r / v  # Periodo del movimiento
omega = 2 * np.pi / T  # Velocidad angular

# Tiempo
t = np.linspace(0, 2 * T, 1000)  # Tiempo desde 0 hasta 2 periodos

# Movimiento circular
x = r * np.cos(omega * t)
y = r * np.sin(omega * t)

# Forma de onda senoidal (proyección en y)
y_senoidal = r * np.sin(omega * t)

# Gráfica del movimiento circular en el plano cartesiano
plt.figure(figsize=(14, 6))

# Gráfica 1: Movimiento circular
plt.subplot(1, 2, 1)
plt.plot(x, y, label='Movimiento Circular', color='blue')
plt.title("Movimiento Circular Uniforme")
plt.xlabel("x (metros)")
plt.ylabel("y (metros)")
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')

# Gráfica 2: Forma de onda senoidal
plt.subplot(1, 2, 2)
plt.plot(t, y_senoidal, label='Forma de Onda Senoidal', color='red')
plt.title("Proyección en el Eje y (Forma de Onda Senoidal)")
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud (m)")
plt.grid(True)

plt.tight_layout()
plt.show()

# Resultados
print(f"{'Parámetro':<25} {'Valor':>10}")
print(f"{'-' * 35}")
print(f"{'Radio (r)':<25} {r:>10.2f} m")
print(f"{'Velocidad Tangencial (v)':<25} {v:>10.2f} m/s")
print(f"{'Periodo (T)':<25} {T:>10.2f} s")
print(f"{'Velocidad Angular (ω)':<25} {omega:>10.2f} rad/s")
```
