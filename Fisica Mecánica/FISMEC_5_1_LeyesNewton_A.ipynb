# Leyes de Newton

## Primera Ley de Newton: Ley de la Inercia

La **Primera Ley de Newton**, también conocida como la *Ley de la Inercia*, establece que:

> **Un objeto permanecerá en reposo o en movimiento rectilíneo uniforme a menos que una fuerza externa neta actúe sobre él.**

Esto significa que si no hay una fuerza neta que actúe sobre un objeto, su estado de movimiento no cambiará. Esta ley nos ayuda a comprender que la inercia es la tendencia de un objeto a resistir cambios en su estado de movimiento.

### Expresión matemática

Aunque la primera ley no se expresa en una fórmula específica, podemos decir que si la fuerza neta es cero, entonces:

$$
\sum \vec{F} = 0 \Rightarrow \vec{v} = \text{constante}
$$

donde:
- $\sum \vec{F}$ es la suma de todas las fuerzas aplicadas en el objeto.
- $\vec{v}$ es la velocidad del objeto, que permanecerá constante si no hay fuerza neta.

### Ejemplos de la Primera Ley de Newton

1. **Uso del cinturón de seguridad:** Cuando un auto en movimiento frena repentinamente, las personas dentro tienden a seguir moviéndose hacia adelante. Esto es debido a la inercia de su cuerpo. El cinturón de seguridad actúa como la fuerza externa que cambia el estado de movimiento y evita que sigan avanzando.
   
2. **Deslizar un libro en una mesa:** Si deslizamos un libro sobre una superficie lisa, eventualmente se detendrá debido a la fricción. Si no hubiera fricción, el libro continuaría en movimiento indefinidamente, como lo describe la primera ley.

---



### Ejemplos de la Primera Ley de Newton con Cálculos

#### Ejemplo 1: Resistencia de fricción en una superficie horizontal

Imagina que empujas un libro de 1 kg sobre una mesa y, después de un tiempo, el libro se detiene debido a la fricción. Queremos calcular la fuerza de fricción que detuvo el libro.

- **Datos conocidos:**
  - Masa del libro, $m = 1 \, \text{kg}$
  - La velocidad inicial es $v_0 = 3 \, \text{m/s}$ (inicialmente empujado)
  - La distancia recorrida hasta detenerse, $d = 2 \, \text{m}$.

Usamos la ecuación de energía para relacionar la energía inicial y la pérdida de energía debido a la fuerza de fricción:

$$
\text{Trabajo de fricción} = \text{Cambio de energía cinética} = \frac{1}{2} m v_0^2
$$

Como la fricción es la única fuerza que realiza trabajo, podemos escribir:

$$
f_{\text{fricción}} \cdot d = \frac{1}{2} m v_0^2
$$

Despejamos $f_{\text{fricción}}$:

$$
f_{\text{fricción}} = \frac{\frac{1}{2} m v_0^2}{d}
$$

- **Sustitución de valores:**

$$
f_{\text{fricción}} = \frac{\frac{1}{2} \cdot 1 \, \text{kg} \cdot (3 \, \text{m/s})^2}{2 \, \text{m}} = 2.25 \, \text{N}
$$

Entonces, la fuerza de fricción que detuvo al libro fue de $2.25 \, \text{N}$.

---

#### Ejemplo 2: Fuerza para vencer la inercia en un plano inclinado sin fricción

Un bloque de $2 \, \text{kg}$ se encuentra en un plano inclinado de $30^\circ$ sin fricción. Queremos encontrar la fuerza que debe aplicarse en dirección contraria a la pendiente para evitar que el bloque empiece a deslizarse.

- **Datos conocidos:**
  - Masa del bloque, $m = 2 \, \text{kg}$
  - Gravedad, $g = 9.81 \, \text{m/s}^2$
  - Ángulo del plano, $\theta = 30^\circ$

El componente de la fuerza de gravedad que actúa sobre el bloque en la dirección del plano es:

$$
f_{\text{gravedad, paralelo}} = m \cdot g \cdot \sin(\theta)
$$

Para que el bloque no se mueva, la fuerza aplicada, $f_{\text{aplicada}}$, debe igualar esta componente de la gravedad.

$$
f_{\text{aplicada}} = m \cdot g \cdot \sin(\theta)
$$

- **Sustitución de valores:**

$$
f_{\text{aplicada}} = 2 \, \text{kg} \cdot 9.81 \, \text{m/s}^2 \cdot \sin(30^\circ) = 9.81 \, \text{N}
$$

Por lo tanto, la fuerza que necesitamos aplicar para mantener el bloque en reposo es de $9.81 \, \text{N}$.

## Segunda Ley de Newton: Ley de la Fuerza y Aceleración

La **segunda ley de Newton** establece que la aceleración de un objeto es directamente proporcional a la fuerza neta que actúa sobre él e inversamente proporcional a su masa. En otras palabras, la fuerza aplicada a un objeto determina su aceleración, dependiendo de la masa del objeto.

### Fórmula de la Segunda Ley de Newton

La segunda ley de Newton se expresa mediante la fórmula:

$$
\vec{F} = m \cdot \vec{a}
$$

donde:

- $\vec{F}$ es la fuerza neta aplicada en newtons (N),
- $m$ es la masa del objeto en kilogramos (kg),
- $\vec{a}$ es la aceleración producida en metros por segundo al cuadrado $(\text{m/s}^2)$.

La dirección de la aceleración es la misma que la de la fuerza neta aplicada.

---

### Ejemplo 1: Cálculo de la Fuerza Necesaria para Acelerar un Cuerpo

Un automóvil de $1,000 \, \text{kg}$ necesita acelerar a $3 \, \text{m/s}^2$ en una autopista. ¿Cuál es la fuerza neta necesaria para lograr esta aceleración?

1. **Datos conocidos**:
   - Masa, $m = 1,000 \, \text{kg}$
   - Aceleración, $a = 3 \, \text{m/s}^2$

2. **Aplicación de la segunda ley**:

   Usamos la fórmula de la segunda ley de Newton:

   $$
   F = m \cdot a
   $$

3. **Sustitución de valores**:

   $$
   F = 1,000 \, \text{kg} \cdot 3 \, \text{m/s}^2 = 3,000 \, \text{N}
   $$

Por lo tanto, la fuerza neta necesaria es de $3,000 \, \text{N}$ en la dirección de la aceleración.

---

### Ejemplo 2: Fuerza Necesaria para Comprimir un Resorte con Masa

Un resorte está conectado a una masa de $5 \, \text{kg}$ en una superficie horizontal sin fricción. Cuando se aplica una fuerza de $15 \, \text{N}$, la masa se desplaza y se comprime el resorte. ¿Cuál es la aceleración producida?

1. **Datos conocidos**:
   - Masa, $m = 5 \, \text{kg}$
   - Fuerza aplicada, $F = 15 \, \text{N}$

2. **Aplicación de la segunda ley**:

   Usamos la fórmula de la segunda ley de Newton y despejamos $a$:

   $$
   a = \frac{F}{m}
   $$

3. **Sustitución de valores**:

   $$
   a = \frac{15 \, \text{N}}{5 \, \text{kg}} = 3 \, \text{m/s}^2
   $$

Por lo tanto, la aceleración producida en la masa al aplicar la fuerza es de $3 \, \text{m/s}^2$.

## Tercera Ley de Newton: Acción y Reacción

La **tercera ley de Newton** establece que **por cada acción, existe una reacción igual y opuesta**. Esto significa que cuando un objeto ejerce una fuerza sobre un segundo objeto, el segundo objeto ejerce una fuerza de igual magnitud pero en dirección opuesta sobre el primero.

### Fórmula de la Tercera Ley de Newton

Si un objeto A ejerce una fuerza $\vec{F}_{A}$ sobre un objeto B, entonces el objeto B ejercerá una fuerza $\vec{F}_{B}$ sobre el objeto A de igual magnitud pero en sentido opuesto:

$$
\vec{F}_{A \rightarrow B} = -\vec{F}_{B \rightarrow A}
$$

donde:

- $\vec{F}_{A \rightarrow B}$ es la fuerza que el objeto A ejerce sobre el objeto B,
- $\vec{F}_{B \rightarrow A}$ es la fuerza que el objeto B ejerce sobre el objeto A.

Estas fuerzas de acción y reacción son siempre iguales en magnitud y opuestas en dirección, y actúan sobre cuerpos diferentes.

---

### Ejemplo: Acción y Reacción entre una Persona y el Suelo

Imagina una persona de $60 \, \text{kg}$ que está de pie sobre el suelo. La gravedad ejerce una fuerza hacia abajo sobre la persona, que es igual a su peso. La persona, a su vez, ejerce una fuerza de igual magnitud hacia abajo sobre el suelo. ¿Qué fuerza ejerce el suelo en respuesta?

1. **Datos conocidos**:
   - Masa de la persona, $m = 60 \, \text{kg}$
   - Aceleración gravitacional, $g = 9.81 \, \text{m/s}^2$

2. **Cálculo del peso (fuerza gravitacional)**:

   La fuerza que la persona ejerce hacia abajo sobre el suelo es:

   $$
   F_{\text{peso}} = m \cdot g = 60 \, \text{kg} \cdot 9.81 \, \text{m/s}^2 = 588.6 \, \text{N}
   $$

3. **Fuerza de reacción del suelo**:

   Según la tercera ley de Newton, el suelo ejerce una fuerza de igual magnitud, pero en dirección opuesta, hacia arriba sobre la persona:

   $$
   F_{\text{suelo}} = - F_{\text{peso}} = -588.6 \, \text{N}
   $$

Por lo tanto, el suelo aplica una fuerza de $588.6 \, \text{N}$ hacia arriba sobre la persona.

## Ejemplo: Cálculo de \( k \) para un Lanzador con Resorte

**Dado:**
- Compresión del resorte,  x = 7cm en metros 0.07 m
- Masa de la bola,  m = 50 g  en kilos 0.05 kg
- Distancia horizontal recorrida,  R = 5 m
- Ángulo de lanzamiento, $ \theta = 45^\circ$
- Gravedad, $ g = 9.81 \, \text{m/s}^2 $

Queremos calcular la constante \( k \) del resorte.

### Paso 1: Determinar la Velocidad Inicial Necesaria (\( v_0 \))

La distancia horizontal en un lanzamiento parabólico está dada por:

$$
R = \frac{v_0^2 \sin(2 \theta)}{g}
$$

Despejamos \( v_0 \):

$$
v_0 = \sqrt{\frac{R \cdot g}{\sin(2 \theta)}}
$$

Sustituyendo los valores:

$$
v_0 = \sqrt{\frac{5 \cdot 9.81}{\sin(90^\circ)}} = \sqrt{5 \cdot 9.81} = \sqrt{49.05} \approx 7 \, \text{m/s}
$$

Entonces, la velocidad inicial necesaria para alcanzar los 5 m es \( v_0 = 7 \, \text{m/s} \).

### Paso 2: Usar la Energía Cinética para Determinar la Energía Almacenada en el Resorte

La energía cinética inicial \( E_k \) de la bolita al ser lanzada es:

$$
E_k = \frac{1}{2} m v_0^2
$$

Sustituyendo:

$$
E_k = \frac{1}{2} \cdot 0.05 \cdot (7)^2 = 0.025 \cdot 49 = 1.225 \, \text{J}
$$

Esta energía cinética proviene de la energía potencial elástica del resorte cuando está comprimido.

### Paso 3: Igualar la Energía Potencial Elástica del Resorte a la Energía Cinética de la Bolita

La energía potencial elástica en el resorte comprimido es:

$$
E_p = \frac{1}{2} k x^2
$$

Igualamos \( E_p \) y \( E_k \):

$$
\frac{1}{2} k x^2 = 1.225
$$

Despejamos \( k \):

$$
k = \frac{2 \cdot 1.225}{x^2} = \frac{2 \cdot 1.225}{(0.07)^2}
$$

Calculamos:

$$
k = \frac{2.45}{0.0049} \approx 500 \, \text{N/m}
$$

---

### Resultado Final
La constante del resorte necesaria para lanzar la bolita 5 metros es aproximadamente:

$$
k \approx 500 \, \text{N/m}
$$
