# Lanzamiento de Proyectiles

## 1. **Teoría Básica del Lanzamiento de Proyectiles**

El **lanzamiento de proyectiles** es un tipo de movimiento en el plano que combina un movimiento horizontal uniforme con un movimiento vertical uniformemente acelerado debido a la gravedad. Este tipo de movimiento es una excelente manera de entender cómo las fuerzas actúan sobre un objeto en dos dimensiones.

### **Componentes del Movimiento**

- **Movimiento Horizontal (Eje X):**
  - Movimiento rectilíneo uniforme (MRU).
  - Velocidad horizontal constante: 
    $$x(t) = v_0 \cos(\theta) \cdot t$$
  
- **Movimiento Vertical (Eje Y):**
  - Movimiento uniformemente acelerado (MUU) con aceleración debida a la gravedad.
  - Desplazamiento vertical: 
    $$y(t) = v_0 \sin(\theta) \cdot t - \frac{1}{2} g t^2$$

### **Parámetros Importantes**

- $ v_0:$ Velocidad inicial del proyectil (m/s).
- $ \theta $: Ángulo de lanzamiento con respecto al eje horizontal (grados o radianes).
- $ g $: Aceleración debida a la gravedad $(9.81 \, \text{m/s}^2$.
- **Tiempo de vuelo**: Tiempo total que el proyectil está en el aire.
- **Altura máxima**: Punto más alto alcanzado por el proyectil.
- **Alcance horizontal**: Distancia total recorrida por el proyectil en el eje horizontal.

## 2. **Ecuaciones del Movimiento**

### **Ecuaciones Paramétricas**

- **Posición Horizontal:**
  $$x(t) = v_0 \cos(\theta) \cdot t$$

- **Posición Vertical:**
  $$y(t) = v_0 \sin(\theta) \cdot t - \frac{1}{2} g t^2$$

### **Cálculos Clave**

- **Tiempo Total de Vuelo:**
  $$t_{\text{vuelo}} = \frac{2 v_0 \sin(\theta)}{g}$$

- **Alcance Máximo:**
  $$R = \frac{v_0^2 \sin(2\theta)}{g}$$

- **Altura Máxima:**
  $$h_{\text{máxima}} = \frac{(v_0 \sin(\theta))^2}{2g}$$

## 3. **Ejemplo Resuelto**

### **Enunciado:**

Un proyectil se lanza con una velocidad inicial de \(20 \, \text{m/s}\) formando un ángulo de \(45^\circ\) con la horizontal. Calcula el tiempo total de vuelo, el alcance máximo y la altura máxima.

### **Solución:**

1. **Datos:**
   - $ v_0 = 20 \, \text{m/s} $
   - $ \theta = 45^\circ $ $ = \frac{\pi}{4} \, \text{rad} $
   - $ g = 9.81 \, \text{m/s}^2 $

2. **Cálculo de Componentes de la Velocidad Inicial:**
   $$v_{0x} = v_0 \cos(\theta) = 20 \cdot \cos\left(\frac{\pi}{4}\right) \approx 14.14 \, \text{m/s}$$
   $$v_{0y} = v_0 \sin(\theta) = 20 \cdot \sin\left(\frac{\pi}{4}\right) \approx 14.14 \, \text{m/s}$$

3. **Tiempo Total de Vuelo:**
   $$t_{\text{vuelo}} = \frac{2 v_{0y}}{g} = \frac{2 \cdot 14.14}{9.81} \approx 2.89 \, \text{s}$$

4. **Alcance Máximo:**
   $$R = v_{0x} \cdot t_{\text{vuelo}} = 14.14 \cdot 2.89 \approx 40.86 \, \text{m}$$

5. **Altura Máxima:**
   $$h_{\text{máxima}} = \frac{v_{0y}^2}{2g} = \frac{(14.14)^2}{2 \cdot 9.81} \approx 10.20 \, \text{m}$$

### **Resultados:**

- **Tiempo Total de Vuelo:** $\approx 2.89 \, \text{s}$
- **Alcance Máximo:** $ \approx 40.86 \, \text{m} $
- **Altura Máxima:** $ \approx 10.20 \, \text{m} $



```python

import numpy as np
import matplotlib.pyplot as plt

# Datos del lanzamiento
v0 = 20  # Velocidad inicial en m/s
theta_deg = 45  # Ángulo de lanzamiento en grados
theta = np.radians(theta_deg)  # Convertir a radianes
g = 9.81  # Aceleración debido a la gravedad en m/s^2

# Componentes de la velocidad inicial
v0x = v0 * np.cos(theta)
v0y = v0 * np.sin(theta)

# Tiempo total de vuelo
t_total = (2 * v0y) / g

# Tiempo en diferentes puntos de la trayectoria
t = np.linspace(0, t_total, num=500)

# Ecuaciones paramétricas de la trayectoria
x = v0x * t
y = v0y * t - 0.5 * g * t**2

# Graficar la trayectoria
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Trayectoria del Proyectil')
plt.title("Trayectoria de un Proyectil")
plt.xlabel("Distancia Horizontal (m)")
plt.ylabel("Altura (m)")
plt.grid(True)
plt.legend()
plt.show()

# Cálculos adicionales
alcance_max = v0x * t_total
altura_max = (v0y**2) / (2 * g)

print(f"Tiempo Total de Vuelo: {t_total:.2f} s")
print(f"Alcance Máximo: {alcance_max:.2f} m")
print(f"Altura Máxima: {altura_max:.2f} m")
```
