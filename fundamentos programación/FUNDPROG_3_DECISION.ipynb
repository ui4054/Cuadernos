## Estructuras de Control - Toma de Desición - IF ELSE

### Condicionales
Las estructuras condicionales permiten ejecutar diferentes bloques de código en función de una condición. En C#, utilizamos `if`, `else if` y `else` para las condicionales.

Ejemplo:

```csharp
int numero = 10;

if (numero > 0)
{
    Console.WriteLine("El número es positivo");
}
else if (numero < 0)
{
    Console.WriteLine("El número es negativo");
}
else
{
    Console.WriteLine("El número es cero");
}

## Ejercicio 1: Determinar si un número es primo

### Enunciado
Escribe un programa que determine si un número ingresado por el usuario es primo o no. Un número primo es aquel que solo es divisible por 1 y por sí mismo.

### Código en C#

```csharp
 Console.Write("Ingrese un número: ");
        int numero = int.Parse(Console.ReadLine());
        bool esPrimo = true;

        if (numero <= 1)
        {
            esPrimo = false;
        }
        else
        {
            for (int i = 2; i <= Math.Sqrt(numero); i++)
            {
                if (numero % i == 0)
                {
                    esPrimo = false;
                    break;
                }
            }
        }

        if (esPrimo)
        {
            Console.WriteLine(numero + " es un número primo.");
        }
        else
        {
            Console.WriteLine(numero + "no es un número primo.");
        }


## Ejercicio 2: Determinar el mayor de tres números

### Enunciado
Escribe un programa que determine el mayor de tres números ingresados por el usuario.

### Código en C#

```csharp

        Console.Write("Ingrese el primer número: ");
        int num1 = int.Parse(Console.ReadLine());

        Console.Write("Ingrese el segundo número: ");
        int num2 = int.Parse(Console.ReadLine());

        Console.Write("Ingrese el tercer número: ");
        int num3 = int.Parse(Console.ReadLine());

        int mayor = num1;

        if (num2 > mayor)
        {
            mayor = num2;
        }

        if (num3 > mayor)
        {
            mayor = num3;
        }

        Console.WriteLine($"El mayor de los tres números es: {mayor}");


## Ejercicio 3: Determinar si un año es bisiesto

### Enunciado
Escribe un programa que determine si un año ingresado por el usuario es bisiesto o no. Un año es bisiesto si es divisible por 4, pero no por 100, a menos que también sea divisible por 400.

### Código en C#

```csharp
using System;


        Console.Write("Ingrese un año: ");
        int año = int.Parse(Console.ReadLine());
        bool esBisiesto = false;

        if (año % 4 == 0)
        {
            if (año % 100 != 0 || año % 400 == 0)
            {
                esBisiesto = true;
            }
        }

        if (esBisiesto)
        {
            Console.WriteLine(año + " es un año bisiesto.");
        }
        else
        {
            Console.WriteLine(año + " no es un año bisiesto.");
        }



## Ejercicio 4: Clasificar una calificación

### Enunciado
Escribe un programa que clasifique una calificación ingresada por el usuario en las siguientes categorías: "A" (90-100), "B" (80-89), "C" (70-79), "D" (60-69), y "F" (0-59).

### Código en C#

```csharp
using System;


        Console.Write("Ingrese una calificación (0-100): ");
        int calificacion = int.Parse(Console.ReadLine());
        string categoria;

        if (calificacion >= 90)
        {
            categoria = "A";
        }
        else if (calificacion >= 80)
        {
            categoria = "B";
        }
        else if (calificacion >= 70)
        {
            categoria = "C";
        }
        else if (calificacion >= 60)
        {
            categoria = "D";
        }
        else
        {
            categoria = "F";
        }

        Console.WriteLine($"La calificación es: {categoria}");




## Ejercicio 5: Descuento en una tienda de ropa

### Enunciado
Escribe un programa que determine el descuento aplicable en una tienda de ropa según los siguientes criterios:
- Si compra ropa de niño y niña, aplica un 10% de descuento.
- Si compra más de una docena de productos, aplica un 15% de descuento.
- Si compra ropa para mujer, aplica un 5% de descuento.
- Si compra ropa para hombre, aplica un 5% de descuento.
- Si no cumple ninguna de las condiciones anteriores, no se aplica descuento.

### Código en C#

```csharp
using System;

        Console.Write("Ingrese la cantidad de ropa de niño: ");
        int ropaNiño = int.Parse(Console.ReadLine());

        Console.Write("Ingrese la cantidad de ropa de niña: ");
        int ropaNiña = int.Parse(Console.ReadLine());

        Console.Write("Ingrese la cantidad de ropa de mujer: ");
        int ropaMujer = int.Parse(Console.ReadLine());

        Console.Write("Ingrese la cantidad de ropa de hombre: ");
        int ropaHombre = int.Parse(Console.ReadLine());

        int totalProductos = ropaNiño + ropaNiña + ropaMujer + ropaHombre;
        double descuento = 0;

        if (ropaNiño > 0 && ropaNiña > 0)
        {
            descuento = 0.10;
        }
        else if (totalProductos > 12)
        {
            descuento = 0.15;
        }
        else if (ropaMujer > 0)
        {
            descuento = 0.05;
        }
        else if (ropaHombre > 0)
        {
            descuento = 0.05;
        }

        Console.WriteLine($"El descuento aplicable es: {descuento * 100}%");



## Instrucción `switch-case` en C#

La instrucción `switch-case` en C# se utiliza para seleccionar una de entre varias secciones de código a ejecutar, basándose en el valor de una expresión. Es una alternativa más organizada y legible a múltiples declaraciones `if-else`.

### Sintaxis

```csharp
switch (expresion)
{
    case valor1:
        // Código a ejecutar si expresion == valor1
        break;
    case valor2:
        // Código a ejecutar si expresion == valor2
        break;
    // Puedes añadir tantos casos como sean necesarios
    default:
        // Código a ejecutar si ningun caso coincide con la expresion
        break;
}

## Ejemplo con `int`: Menú de opciones

#### Explicación paso a paso:
1. Definimos una variable `día` de tipo `int` que contiene el nombre de un dia de la semana.
2. Utilizamos la instrucción `switch-case` para imprimir una descripción del dia de la semana correspondiente.

```csharp
using System;


        // Definir un número que representa el día de la semana (1 para Lunes, 2 para Martes, etc.)
        int dia = 3; // Puedes cambiar este valor para probar diferentes días

        // Utilizar la instrucción switch-case para determinar el día de la semana
        switch (dia)
        {
            case 1:
                Console.WriteLine("Lunes");
                break;
            case 2:
                Console.WriteLine("Martes");
                break;
            case 3:
                Console.WriteLine("Miércoles");
                break;
            case 4:
                Console.WriteLine("Jueves");
                break;
            case 5:
                Console.WriteLine("Viernes");
                break;
            case 6:
                Console.WriteLine("Sábado");
                break;
            case 7:
                Console.WriteLine("Domingo");
                break;
            default:
                Console.WriteLine("Número inválido");
                break;
        }
 

## Ejemplo con `char`: Menú de opciones

#### Explicación paso a paso:
1. Definimos una variable `opcion` de tipo `char` que contiene el nombre de una opción.
2. Utilizamos la instrucción `switch-case` para imprimir una descripción de la opción correspondiente.

```csharp
using System;


        // Definir una variable de tipo char para representar la opción seleccionada
        char opcion = 'b'; // Puedes cambiar este valor para probar diferentes opciones

        // Utilizar la instrucción switch-case para ejecutar el código correspondiente a cada opción
        switch (opcion)
        {
            case 'a':
                Console.WriteLine("Opción A seleccionada.");
                break;
            case 'b':
                Console.WriteLine("Opción B seleccionada.");
                break;
            case 'c':
                Console.WriteLine("Opción C seleccionada.");
                break;
            default:
                Console.WriteLine("Opción no válida.");
                break;
        }


## Ejemplo con `string`: Determinar la estación del año

#### Explicación paso a paso:
1. Definimos una variable `estación` de tipo `string` que contiene el nombre de una estación del año.
2. Utilizamos la instrucción `switch-case` para imprimir una descripción de la estación correspondiente.



```csharp
using System;


        // Definir una variable de tipo string para representar la estación del año
        string estacion = "Verano"; // Puedes cambiar este valor para probar diferentes estaciones

        // Utilizar la instrucción switch-case para determinar la estación del año
        switch (estacion)
        {
            case "Primavera":
                Console.WriteLine("Es Primavera: las flores florecen.");
                break;
            case "Verano":
                Console.WriteLine("Es Verano: hace calor.");
                break;
            case "Otoño":
                Console.WriteLine("Es Otoño: las hojas caen.");
                break;
            case "Invierno":
                Console.WriteLine("Es Invierno: hace frío.");
                break;
            default:
                Console.WriteLine("Estación no válida.");
                break;
        }
