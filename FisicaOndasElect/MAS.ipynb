# Cinemática del Movimiento Armónico

## Definición del Movimiento Armónico Simple (MAS)

El **Movimiento Armónico Simple (MAS)** es un tipo de movimiento oscilatorio en el que una partícula se mueve a lo largo de una trayectoria lineal, siguiendo una función sinusoidal del tiempo. Es uno de los ejemplos más simples de movimiento vibratorio y se describe mediante una ecuación diferencial lineal.

## Componentes del MAS

El MAS se caracteriza por varios parámetros clave:

- **Amplitud (A):** Es el valor máximo del desplazamiento respecto al equilibrio. Representa el "tamaño" de la oscilación.
- **Frecuencia (f):** Es el número de oscilaciones completas realizadas en una unidad de tiempo. Se mide en Hertz (Hz).
- **Periodo (T):** Es el tiempo que tarda el sistema en realizar una oscilación completa. Está relacionado con la frecuencia por la fórmula \( T = \frac{1}{f} \).
- **Fase (φ):** Es el ángulo en el que empieza la oscilación en relación al origen. Determina el punto en el ciclo en el que se encuentra el sistema en un instante dado.

## Gráficas del Desplazamiento, Velocidad y Aceleración

Las gráficas del MAS son sinusoidales y pueden representarse como:

- **Desplazamiento (x(t)):** La posición de la partícula en función del tiempo.
- **Velocidad (v(t)):** La tasa de cambio del desplazamiento con respecto al tiempo.
- **Aceleración (a(t)):** La tasa de cambio de la velocidad con respecto al tiempo.

Estas gráficas pueden ser representadas por las siguientes funciones:

- **Desplazamiento:** $$( x(t) = A \cos(\omega t + \phi) )$$
- **Velocidad:** 
  La velocidad es la derivada del desplazamiento con respecto al tiempo:

  $$[
  v(t) = \frac{dx(t)}{dt} = -A \omega \sin(\omega t + \phi)
  ]$$
  $$( v(t) = -A \omega \sin(\omega t + \phi) )$$

- **Aceleración:** 
  La aceleración es la derivada de la velocidad con respecto al tiempo:
  $$[
  a(t) = \frac{dv(t)}{dt} = -A \omega^2 \cos(\omega t + \phi)
  ]$$
  $$( a(t) = -A \omega^2 \cos(\omega t + \phi) )$$

Donde $$( \omega = 2 \pi f )$$ es la frecuencia angular.

## Ejemplos de Sistemas que Exhiben MAS

Algunos ejemplos de sistemas que exhiben MAS incluyen:

- Un péndulo simple con amplitudes pequeñas.
- Un resorte con una masa colgante.
- Las vibraciones de una cuerda de guitarra cuando se pulsa y se suelta.

## Ejemplo 1: Cálculo de la Frecuencia de un Péndulo Simple

Consideremos un péndulo simple con una longitud \( L \) y una masa \( m \). La frecuencia de oscilación del péndulo simple está relacionada con la longitud del péndulo y la aceleración debida a la gravedad \( g \). 

### Fórmula para la Frecuencia del Péndulo Simple

La frecuencia angular $$( \omega )$$de un péndulo simple se calcula mediante la fórmula:

$$
\omega = \sqrt{\frac{g}{L}}
$$

Donde:
- \( g \) es la aceleración debida a la gravedad (aproximadamente \( 9.81 \, \text{m/s}^2 \)).
- \( L \) es la longitud del péndulo en metros.

La frecuencia \( f \) en Hertz (Hz) se relaciona con la frecuencia angular mediante la fórmula:

$$
f = \frac{\omega}{2 \pi}
$$

### Datos del Problema

Supongamos que tenemos un péndulo simple con:
- Longitud $$( L = 30 \, \text{cm} = 0.30 \, \text{m} )$$
- Masa $$( m = 10 \, \text{g} )$$ (la masa no afecta la frecuencia en un péndulo simple, pero la incluimos para referencia)

### Cálculo

1. **Calcular la Frecuencia Angular \( \omega \):**

   $$
   \omega = \sqrt{\frac{g}{L}} = \sqrt{\frac{9.81}{0.30}} \approx \sqrt{32.7} \approx 5.72 \, \text{rad/s}
   $$

2. **Calcular la Frecuencia \( f \):**

   $$
   f = \frac{\omega}{2 \pi} = \frac{5.72}{2 \pi} \approx \frac{5.72}{6.28} \approx 0.91 \, \text{Hz}
   $$

### Resultado

La frecuencia de oscilación del péndulo simple con una longitud de 30 cm es aproximadamente ( 0.91 {Hz} ).

Este ejemplo muestra cómo la frecuencia del péndulo simple depende únicamente de su longitud y no de la masa del péndulo.

## Ejemplo 2: Cálculo de la Frecuencia de un Resorte en Posición Vertical

Consideremos un sistema masa-resorte en posición vertical. La constante del resorte se puede determinar a partir del peso de la masa que cuelga del resorte. La frecuencia de oscilación del sistema masa-resorte está relacionada con la constante del resorte y la masa.

### Cálculo del Coeficiente de Elasticidad del Resorte

Para calcular el coeficiente de elasticidad del resorte, primero necesitamos usar la ley de Hooke, que establece que la fuerza \( F \) ejercida por un resorte es proporcional a su elongación \( x \):

$$
F = kx
$$

Donde:
- \( F \) es la fuerza aplicada al resorte.
- \( k \) es la constante del resorte.
- \( x \) es la elongación del resorte desde su posición de equilibrio.

En un resorte vertical, la fuerza \( F \) es el peso de la masa colgada, calculado como:

$$
F = m \cdot g
$$

Donde:
- \( m \) es la masa en kilogramos.
- \( g \) es la aceleración debida a la gravedad (aproximadamente \( 9.81 \, \text{m/s}^2 \)).

Podemos despejar \( k \) usando:

$$
k = \frac{F}{x} = \frac{m \cdot g}{x}
$$

### Fórmula para la Frecuencia del Sistema Masa-Resorte

La frecuencia angular \( \omega \) de un sistema masa-resorte se calcula mediante la fórmula:

$$
\omega = \sqrt{\frac{k}{m}}
$$

Donde:
- \( k \) es la constante del resorte.
- \( m \) es la masa en kilogramos.

La frecuencia \( f \) en Hertz (Hz) se relaciona con la frecuencia angular mediante la fórmula:

$$
f = \frac{\omega}{2 \pi}
$$

### Datos del Problema

Supongamos que tenemos un resorte con una elongación de  0.05 m debido a una masa colgada. La masa colgada es de 200g o 0.2 kg.

### Cálculo del Coeficiente de Elasticidad \( k \)

1. **Calcular la Fuerza Aplicada \( F \):**

   $$
   F = m \cdot g = 0.2 \, \text{kg} \times 9.81 \, \text{m/s}^2 = 1.962 \, \text{N}
   $$

2. **Calcular \( k \):**

   $$
   k = \frac{F}{x} = \frac{1.962}{0.05} = 39.24 \, \text{N/m}
   $$

### Cálculo de la Frecuencia del Sistema Masa-Resorte

1. **Calcular la Frecuencia Angular (w):**

   $$
   \omega = \sqrt{\frac{k}{m}} = \sqrt{\frac{39.24}{0.2}} \approx \sqrt{196.2} \approx 14.00 \, \text{rad/s}
   $$

2. **Calcular la Frecuencia \( f \):**

   $$
   f = \frac{\omega}{2 \pi} = \frac{14.00}{2 \pi} \approx \frac{14.00}{6.28} \approx 2.23 \, \text{Hz}
   $$

### Resultado

La frecuencia de oscilación del sistema masa-resorte con una elongación de 0.05 m debido a una masa de 200 g es aproximadamente \( 2.23 \, \text{Hz} \).

Este ejemplo muestra cómo calcular la constante del resorte a partir del peso de la masa y la elongación, y luego usar esa constante para determinar la frecuencia de oscilación del sistema masa-resorte en posición vertical.

# Energía en un Oscilador Armónico Simple (OAS)

## Introducción

En un oscilador armónico simple (OAS), la energía total se conserva y se distribuye entre la energía cinética y la energía potencial a medida que el sistema oscila. Este cuaderno explora cómo se comportan estas energías y su relación con la posición y la velocidad del oscilador.

## Energía Cinética \(E_k\)

La energía cinética de un oscilador armónico simple está dada por:

$$
E_k = \frac{1}{2}mv^2
$$

Donde:
- \( m \) es la masa del objeto.
- \( v \) es la velocidad, que se expresa como:

$$
v(t) = -A\omega \sin(\omega t)
$$

Sustituyendo esta expresión en la fórmula para \(E_k\):

$$
E_k(t) = \frac{1}{2}m(A\omega \sin(\omega t))^2
$$

## Energía Potencial \(E_p\)

La energía potencial está dada por:

$$
E_p = \frac{1}{2}kx^2
$$

Donde:
- \( k \) es la constante del sistema.
- \( x \) es la posición, que se expresa como:

$$
x(t) = A\cos(\omega t)
$$

Sustituyendo esta expresión en la fórmula para \(E_p\):

$$
E_p(t) = \frac{1}{2}k(A\cos(\omega t))^2
$$

## Energía Total \(E_t\)

La energía total en el oscilador es la suma de la energía cinética y la energía potencial:

$$
E_t = E_k + E_p
$$

En un sistema conservativo, la energía total se mantiene constante y está dada por:

$$
E_t = \frac{1}{2}kA^2 = \frac{1}{2}m\omega^2A^2
$$

## Comportamiento de la Energía en el OAS

- **En la posición de equilibrio (\( x = 0 \))**: Toda la energía es cinética.
- **En los extremos (\( x = \pm A \))**: Toda la energía es potencial.
- **En posiciones intermedias**: La energía se distribuye entre \( E_k \) y \( E_p \).

## Ejemplo Numérico

Consideremos un oscilador con los siguientes valores:
- Amplitud \( A = 0.1 \, \text{m} \)
- Masa \( m = 0.5 \, \text{kg} \)
- Constante del resorte \( k = 20 \, \text{N/m} \)

La frecuencia angular \( \omega \) se calcula como:

$$
\omega = \sqrt{\frac{k}{m}} = \sqrt{\frac{20}{0.5}} = \sqrt{40} \approx 6.32 \, \text{rad/s}
$$

La energía total es:

$$
E_t = \frac{1}{2}kA^2 = \frac{1}{2} \times 20 \times (0.1)^2 = 0.1 \, \text{J}
$$

## Visualización en Python

A continuación, generaremos gráficas para mostrar cómo varían \( E_k \), \( E_p \) y \( E_t \) con el tiempo.



```python

import numpy as np
import matplotlib.pyplot as plt

# Parámetros
A = 0.1 # Amplitud en metros
m = 0.5  # Masa en kg
k = 3   # Constante del resorte en N/m
omega = np.sqrt(k / m)  # Frecuencia angular en rad/s
t = np.linspace(0, 2 , 1000)  # Tiempo desde 0 a 10 segundos
#t = 0
# Energías
Ek = 0.5 * m * (A * omega * np.sin(omega * t))**2  # Energía cinética
Ep = 0.5 * k * (A * np.cos(omega * t))**2  # Energía potencial
Et = Ek + Ep  # Energía total (constante)

# Gráficas
plt.figure(figsize=(10, 6))
plt.plot(t, Ek, label='Energía Cinética \(E_k\)')
plt.plot(t, Ep, label='Energía Potencial \(E_p\)')
plt.plot(t, Et, label='Energía Total \(E_t\)', linestyle='--')
plt.xlabel('Tiempo (s)')
plt.ylabel('Energía (J)')
plt.title('Energía en un Oscilador Armónico Simple')
plt.legend()
plt.grid(True)
plt.show()
```

```python
### oscilador armionico en pendulo.
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
l = 0.2 # longitud de la cuerda
m = 0.5 # masa puntual de objeto
I = m*l*l  # Momento de Inercia rotacion
g = 9.8   # Constante gravitacional
omega = np.sqrt(g / l)  # Frecuencia angular en rad/s
A = 0.1 *(1-np.cos(omega))  # Amplitud en metros
t = np.linspace(0, 2 , 1000)  # Tiempo desde 0 a 10 segundos
#t = 0
# Energías
Ek = 0.5 * I * (A * omega * np.sin(omega * t))**2  # Energía cinética
Ep = 0.5 *m *g * l* (A * np.cos(omega * t))**2  # Energía potencial
Et = Ek + Ep  # Energía total (constante)

# Gráficas
plt.figure(figsize=(10, 6))
plt.plot(t, Ek, label='Energía Cinética \(E_k\)')
plt.plot(t, Ep, label='Energía Potencial \(E_p\)')
plt.plot(t, Et, label='Energía Total \(E_t\)', linestyle='--')
plt.xlabel('Tiempo (s)')
plt.ylabel('Energía (J)')
plt.title('Energía en un Oscilador Armónico Simple')
plt.legend()
plt.grid(True)
plt.show()
```
