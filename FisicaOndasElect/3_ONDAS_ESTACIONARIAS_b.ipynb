# Ondas Estacionarias

## 1. Definición

Una **onda estacionaria** es una onda que permanece fija en el espacio, sin propagarse a través del medio. Se forma mediante la superposición de dos ondas progresivas de la misma frecuencia, amplitud y longitud de onda, pero que viajan en direcciones opuestas. Las ondas estacionarias son caracterizadas por la presencia de **nodos** y **antinodos**:

- **Nodos**: Puntos donde el desplazamiento de la onda es siempre cero. En estos puntos, las dos ondas progresivas se cancelan completamente.
- **Antinodos**: Puntos donde el desplazamiento de la onda alcanza su valor máximo. En estos puntos, las dos ondas progresivas se suman para producir una amplitud máxima.

## 2. Ecuación de Ondas Estacionarias

La forma general de una onda estacionaria en una cuerda fija en ambos extremos es:

$$
y(x, t) = 2A \sin(kx) \cos(\omega t)
$$

Donde:
- \( y(x, t) \) es el desplazamiento en la posición \( x \) y tiempo \( t \).
- \( A \) es la amplitud de las ondas progresivas.
- \( k \) es el número de onda, definido como $ k = \frac{2\pi}{\lambda}$, donde $ \lambda $ es la longitud de onda.
- $ \omega $ es la frecuencia angular, definida como $ \omega = 2\pi f $, donde $ f $ es la frecuencia de la onda.

### Derivación de la Ecuación

Las ondas estacionarias se forman por la superposición de dos ondas progresivas:
$$[
y_1(x, t) = A \sin(kx - \omega t)
$$
$$
y_2(x, t) = A \sin(kx + \omega t)
$$

La onda estacionaria es la suma de estas dos ondas:

$$
y(x, t) = y_1(x, t) + y_2(x, t) = A \sin(kx - \omega t) + A \sin(kx + \omega t)
$$

Utilizando la identidad trigonométrica:

$$
\sin \alpha + \sin \beta = 2 \sin\left(\frac{\alpha + \beta}{2}\right) \cos\left(\frac{\alpha - \beta}{2}\right)
$$

Aplicamos esto a la suma de las dos ondas:

$$
y(x, t) = 2A \sin(kx) \cos(\omega t)
$$

## **Recordar Velocidad de la Onda en una Cuerda**

La velocidad de la onda \(v\) en una cuerda se calcula con la siguiente ecuación:

$$
v = \sqrt{\frac{T}{\mu}}
$$

Donde:
- \(T\) es la tensión de la cuerda (en Newtons, N).
- \(\mu\) es la densidad lineal de la cuerda (en kg/m).

### **Descomposición de la Tensión \(T\)**

La **tensión** \(T\) puede ser causada por un peso colgando de la cuerda:

$$
T = m_p \cdot g
$$

Donde:
- \(m_p\) es la masa que cuelga de la cuerda (en kilogramos, kg).
- \(g\) es la aceleración de la gravedad $9.81 \, \text{m/s}^2\ $.

### **Descomposición de la Densidad Lineal \( \mu \)**

La **densidad lineal** de la cuerda es:

$$
\mu = \frac{m_c}{L}
$$

Donde:
- \(m_c\) es la masa de la cuerda (en kilogramos, kg).
- \(L\) es la longitud de la cuerda (en metros, m).

### **Velocidad de la Onda en Función de Componentes**

Sustituyendo las expresiones de \(T\) y \(\mu\):

$$
v = \sqrt{\frac{m_p \cdot g}{\frac{m_c}{L}}}
$$

Simplificando:

$$
v = \sqrt{\frac{m_p \cdot g \cdot L}{m_c}}
$$

### **Ecuación Final**

La velocidad de la onda en función de la masa colgante \(m_p\), la masa de la cuerda \(m_c\), la longitud \(L\) y la gravedad \(g\) es:

$$
v = \sqrt{\frac{m_p \cdot g \cdot L}{m_c}}
$$

### **Interpretación Física**

- **$m_p \cdot g $**: La fuerza de tensión causada por la masa colgante.
- **$ \frac{m_c}{L}$**: La densidad lineal de la cuerda (masa por unidad de longitud).
- **$ v $**: La velocidad de la onda depende de la tensión (peso colgante) y de la densidad de la cuerda.

## 3. Condiciones de Resonancia

Para que se formen ondas estacionarias en una cuerda fija en ambos extremos, deben cumplirse ciertas condiciones de resonancia. Estas condiciones dependen de la longitud de la cuerda y de la longitud de onda de las ondas que se superponen. Las condiciones principales son:

1. **Longitud de la cuerda**:

   La longitud \( L \) de la cuerda debe ser un múltiplo entero de la mitad de la longitud de onda $ \lambda$. Matemáticamente, esto se expresa como:

   $$
   L = n \frac{\lambda}{2} \quad \text{donde} \quad n = 1, 2, 3, \dots
   $$

   Donde \( n \) es el número de nodos (o el modo de vibración).

3. **Frecuencias resonantes**:

   Solo ciertas frecuencias específicas permiten la formación de ondas estacionarias en la cuerda. Estas frecuencias están determinadas por la longitud de la cuerda y la velocidad de la onda \( v \) en el medio:

   $$
   f_n = \frac{n v}{2L} \quad \text{donde} \quad n = 1, 2, 3, \dots
   $$

   Donde $( f_n \)$ es la frecuencia del modo \( n \).

## 4. Ejemplo: Ondas Estacionarias en un Cordón de Zapato

Consideremos un **cordón de zapato** de 2 metros de longitud, fijado en ambos extremos. Supongamos que deseamos encontrar las frecuencias resonantes para los primeros tres modos de vibración.

### Datos:
- **Longitud del cordón (\( L \))**: 2 m
- **Velocidad de la onda (\( v \))**: 22.14 m/s (calculada previamente)
- **Número de nodos (\( n \))**: 1, 2, 3

### Cálculos:

1. **Frecuencia del primer modo (\( n = 1 \))**:

   $$
   f_1 = \frac{1 \times 22.14}{2 \times 2} = \frac{22.14}{4} = 5.535 \, \text{Hz}
   $$

2. **Frecuencia del segundo modo (\( n = 2 \))**:

   $$
   f_2 = \frac{2 \times 22.14}{2 \times 2} = \frac{44.28}{4} = 11.07 \, \text{Hz}
   $$

3. **Frecuencia del tercer modo (\( n = 3 \))**:

   $$
   f_3 = \frac{3 \times 22.14}{2 \times 2} = \frac{66.42}{4} = 16.605 \, \text{Hz}
   $$

### Interpretación:

Estas frecuencias son las **frecuencias resonantes** en las cuales se forman ondas estacionarias en el cordón de zapato. Al vibrar a estas frecuencias, el cordón presenta patrones claros de nodos y antinodos.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import plotly.graph_objects as go

# Parámetros de la cuerda
L = 2.0  # Longitud de la cuerda en metros
A = 0.2  # Amplitud de las ondas en metros
v = 22.14  # Velocidad de la onda en m/s
n = 4 # Número de modo de vibración

# Cálculo de parámetros derivados
lambda_n = (2 * L) / n  # Longitud de onda
k = 2 * np.pi / lambda_n  # Número de onda
f_n = n * v / (2 * L)  # Frecuencia resonante
omega = 2 * np.pi * f_n  # Frecuencia angular

# Tiempo y espacio para la gráfica
x = np.linspace(0, L, 500)  # Posición a lo largo de la cuerda
t = np.linspace(0, 1, 500)  # Tiempo

# Crear malla de espacio y tiempo
X, T_grid = np.meshgrid(x, t)
Y = 2 * A * np.sin(k * X) * np.cos(omega * T_grid)  # Onda estacionaria

# Número de modos a considerar
n_max = 5

# Cálculo de las frecuencias resonantes
frecuencias = [ (n * v) / (2 * L) for n in range(1, n_max + 1) ]

# Crear una tabla de frecuencias
tabla_frecuencias = pd.DataFrame({
    'Modo (n)': range(1, n_max + 1),
    'Frecuencia (Hz)': frecuencias
})


print(tabla_frecuencias)
print('Frecuencia Resonante Activa en Hz',f_n)

# Graficar la onda estacionaria en función de la posición y el tiempo (3D con matplotlib)
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, T_grid, Y, cmap='viridis')
ax.set_title('Onda Estacionaria: Desplazamiento vs Posición y Tiempo')
ax.set_xlabel('Posición (m)')
ax.set_ylabel('Tiempo (s)')
ax.set_zlabel('Desplazamiento (m)')
plt.show()

# Graficar patrones de nodos y antinodos en 2D
plt.figure(figsize=(10, 6))
plt.plot(x, 2 * A * np.sin(k * x), label='Posición fija en t=0')
plt.title('Patrones de Nodos y Antinodos')
plt.xlabel('Posición (m)')
plt.ylabel('Desplazamiento (m)')
plt.axhline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.show()
```
