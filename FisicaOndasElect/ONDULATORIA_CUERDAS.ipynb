# Mecánica Ondulatoria

## 1. Propagación de una perturbación

El fenómeno de **propagación de una perturbación** ocurre cuando una variación en un medio se desplaza a través de él. Ejemplos comunes son las ondas sonoras, las ondas en una cuerda, o las ondas de agua. Una perturbación puede ser cualquier cambio físico, como un desplazamiento, velocidad o presión.

Las características clave de una perturbación que se propaga incluyen:

- **Amplitud**: La máxima variación de la magnitud de la perturbación desde su valor de equilibrio.
- **Frecuencia** ($f$): Número de oscilaciones por unidad de tiempo.
- **Longitud de onda** ($\lambda$): Distancia entre dos puntos equivalentes de ondas consecutivas.
- **Velocidad de propagación** ($v$): Velocidad a la cual la perturbación se desplaza a través del medio, dada por:

$$
v = \lambda f
$$

---

## 2. El modelo de onda progresiva

Una **onda progresiva** es una perturbación que se propaga en un medio en una dirección definida, transportando energía sin que haya un desplazamiento neto de partículas.

El modelo matemático de una onda progresiva, en una dimensión, puede representarse por la siguiente expresión para una onda sinusoidal:

$$
y(x,t) = A \sin(kx - \omega t + \phi)
$$

Donde:
- $A$ es la amplitud de la onda,
- $k = \frac{2\pi}{\lambda}$ es el número de onda,
- $\omega = 2\pi f$ es la frecuencia angular,
- $\phi$ es la fase inicial.

---

## 3. La ecuación de onda lineal

La **ecuación de onda** es una ecuación diferencial parcial que describe la propagación de ondas. Para una onda en una cuerda tensada, la ecuación de onda en una dimensión es:

$$
\frac{\partial^2 y}{\partial x^2} = \frac{1}{v^2} \frac{\partial^2 y}{\partial t^2}
$$

Esta ecuación describe cómo el desplazamiento $y(x,t)$ varía tanto en el espacio ($x$) como en el tiempo ($t$).

---

## 4. La rapidez de ondas en cuerdas

La **rapidez de las ondas** que se propagan en una cuerda depende de la tensión en la cuerda y de la densidad lineal de masa de la cuerda. La fórmula es:

$$
v = \sqrt{\frac{T}{\mu}}
$$

Donde:
- $T$ es la tensión en la cuerda,
- $\mu$ es la densidad lineal de masa de la cuerda (masa por unidad de longitud).

---

## 5. Energía mediante ondas sinusoidales en cuerdas

La **energía** transportada por una onda sinusoidal en una cuerda se asocia con la energía cinética y la energía potencial de las partículas que oscilan. La energía total por unidad de longitud transportada por una onda es:

$$
E = \frac{1}{2} \mu A^2 \omega^2 v
$$

---

## Ejemplos resueltos

### Ejemplo 1: Rapidez de una onda en una cuerda

Una cuerda tiene una tensión de $T = 200 \, \text{N}$ y una densidad lineal de masa de $\mu = 0.01 \, \text{kg/m}$. ¿Cuál es la rapidez de propagación de una onda a través de la cuerda?

**Solución**:

Utilizamos la fórmula de la rapidez de onda:

$$
v = \sqrt{\frac{T}{\mu}}
$$

Sustituyendo los valores:

$$
v = \sqrt{\frac{200 \, \text{N}}{0.01 \, \text{kg/m}}} = \sqrt{20000} = 141.42 \, \text{m/s}
$$

---

### Ejemplo 2: Energía transportada por una onda

Una cuerda con densidad lineal $\mu = 0.02 \, \text{kg/m}$ tiene una onda sinusoidal de amplitud $A = 0.05 \, \text{m}$ y frecuencia angular $\omega = 20 \, \text{rad/s}$. La velocidad de la onda es de $v = 120 \, \text{m/s}$. ¿Cuál es la energía transportada por la onda?

**Solución**:

Usamos la fórmula de la energía total transportada por la onda:

$$
E = \frac{1}{2} \mu A^2 \omega^2 v
$$

Sustituyendo los valores:

$$
E = \frac{1}{2} (0.02) (0.05)^2 (20)^2 (120) = \frac{1}{2} (0.02) (0.0025) (400) (120) = 0.6 \, \text{J}
$$

# Ejemplo de Cálculo de Parámetros de una Onda en un Cordón de Zapato

**Descripción del Problema:**

Se tiene un cordón de zapato atado horizontalmente a un punto fijo en un extremo, y en el otro extremo se cuelga un peso de 500 g. Se desea calcular varios parámetros de la onda que se propaga a lo largo del cordón, incluyendo la densidad lineal, la velocidad de la onda, la frecuencia, el número de onda y el modelo de onda progresiva.

**Datos:**
- Masa del peso: 500 g = 0.5 kg
- Longitud del cordón: $L = 2 \text{ m}$
- Amplitud de la perturbación: 20 cm = 0.2 m
- Masa del cordón: 0.02 kg (suponemos)
- Longitud de onda: $\lambda = 1 \text{ m}$

### 1. Cálculo de la Densidad Lineal

La densidad lineal ($\mu$) se calcula usando la fórmula:

$$
\mu = \frac{\text{masa del cordón}}{\text{longitud}}
$$

En este caso:

$$
\mu = \frac{0.02 \text{ kg}}{2 \text{ m}} = 0.01 \text{ kg/m}
$$

### 2. Cálculo de la Tensión

La tensión ($T$) en el cordón es igual al peso colgado:

$$
T = m \cdot g
$$

Donde $m$ es la masa del peso y $g$ es la aceleración debida a la gravedad ($9.81 \text{ m/s}^2$):

$$
T = 0.5 \text{ kg} \cdot 9.81 \text{ m/s}^2 = 4.905 \text{ N}
$$

### 3. Cálculo de la Velocidad de la Onda

La velocidad de la onda ($v$) se calcula usando:

$$
v = \sqrt{\frac{T}{\mu}}
$$

Sustituyendo los valores:

$$
v = \sqrt{\frac{4.905 \text{ N}}{0.01 \text{ kg/m}}} \approx 22.14 \text{ m/s}
$$

### 4. Cálculo de la Frecuencia y Número de Onda

La frecuencia ($f$) se calcula usando:

$$
f = \frac{v}{\lambda}
$$

Y el número de onda ($k$) se calcula usando:

$$
k = \frac{2\pi}{\lambda}
$$

Para una longitud de onda de $1 \text{ m}$:

$$
f = \frac{22.14 \text{ m/s}}{1 \text{ m}} = 22.14 \text{ Hz}
$$

$$
k = \frac{2\pi}{1 \text{ m}} \approx 6.28 \text{ rad/m}
$$

### 5. Modelo de Onda Progresiva

El modelo de onda progresiva se describe mediante la ecuación:

$$
y(x, t) = A \sin(kx - \omega t)
$$

Donde:
- $A$ es la amplitud de la perturbación ($0.2 \text{ m}$).
- $k$ es el número de onda ($6.28 \text{ rad/m}$).
- $\omega$ es la frecuencia angular, calculada como $\omega = 2\pi f$.

Entonces:

$$
\omega = 2\pi \times 22.14 \text{ Hz} \approx 139.6 \text{ rad/s}
$$

## TALLER - PROYECTO A Energía en Ondas Progresivas

Para calcular la energía transportada por una onda progresiva en un cordón, utilizamos las siguientes fórmulas:

La energía por unidad de longitud de la onda es dada por:

$$
E_{onda} = \frac{1}{2} \mu \omega^2 A^2
$$

Donde:
- $\mu$ es la densidad lineal del cordón (kg/m).
- $\omega$ es la frecuencia angular de la onda (rad/s).
- $A$ es la amplitud de la onda (m).

La energía total transportada por la onda en la cuerda es:

$$
E_{total} = E_{onda} \cdot L
$$

Donde $L$ es la longitud del cordón.

### Ejemplo

Consideramos un cordón de zapato de 2 metros de longitud con una masa de 500 g colgando en un extremo y una amplitud de onda de 0.2 metros. 

**Cálculo de los parámetros:**

- **Tensión en el cordón:** 
$$
T = m_{peso} \cdot g
$$

- **Densidad lineal:**
$$
\mu = \frac{masa_{cordon}}{L}
$$

- **Velocidad de la onda:**
$$
v = \sqrt{\frac{T}{\mu}}
$$

- **Frecuencia:**
$$
f = \frac{v}{longitud_{onda}}
$$

- **Número de onda:**
$$
k = \frac{2 \pi}{longitud_{onda}}
$$

- **Frecuencia angular:**
$$
\omega = 2 \pi f
$$

- **Energía por unidad de longitud:**
$$
E_{onda} = \frac{1}{2} \mu \omega^2 A^2
$$

- **Energía total en el cordón:**
$$
E_{total} = E_{onda} \cdot L
$$

**Código Python:**

```python
import numpy as np

# Datos
m_peso = 0.5  # kg
g = 9.81  # m/s^2
L = 2  # m
A = 0.2  # m
masa_cordon = 0.02  # kg (suponemos)
longitud_onda = 1  # m

# Cálculo de la tensión
T = m_peso * g

# Cálculo de la densidad lineal
mu = masa_cordon / L

# Cálculo de la velocidad de la onda
v = np.sqrt(T / mu)

# Cálculo de la frecuencia
f = v / longitud_onda

# Cálculo del número de onda
k = 2 * np.pi / longitud_onda

# Cálculo de la frecuencia angular
omega = 2 * np.pi * f

# Cálculo de la energía por unidad de longitud
E_onda = 0.5 * mu * omega**2 * A**2

# Cálculo de la energía total
E_total = E_onda * L

# Resultados
print(f"Tensión en el cordón: {T:.2f} N")
print(f"Densidad lineal del cordón: {mu:.2f} kg/m")
print(f"Velocidad de la onda: {v:.2f} m/s")
print(f"Frecuencia de la onda: {f:.2f} Hz")
print(f"Número de onda: {k:.2f} rad/m")
print(f"Frecuencia angular: {omega:.2f} rad/s")
print(f"Energía por unidad de longitud: {E_onda:.2f} J/m")
print(f"Energía total en el cordón: {E_total:.2f} J")

````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import plotly.graph_objects as go

# Datos
m_peso = 0.05  # kg
g = 9.81  # m/s^2
L = 1 # m
A = 0.05  # m
masa_cordon = 0.050 # kg (suponemos)
longitud_onda = 2  # m

# Cálculo de la tensión
T = m_peso * g
ten = T

# Cálculo de la densidad lineal
mu = masa_cordon / L

# Cálculo de la velocidad de la onda
v = np.sqrt(T / mu)

# Cálculo de la frecuencia
f = v / longitud_onda

# Cálculo del número de onda
k = 2 * np.pi / longitud_onda

# Cálculo de la frecuencia angular
omega = 2 * np.pi * f

# Tiempo y espacio para la gráfica
x = np.linspace(0, 2 * longitud_onda, 1000)  # Dos períodos en el espacio
t = np.linspace(0, 2 * (1 / f), 1000)  # Dos períodos en el tiempo

# Crear malla para gráfica 3D
X, T_grid = np.meshgrid(x, t)
Y = A * np.sin(k * X - omega * T_grid)

# Graficar la onda en función de la posición y el tiempo (3D con matplotlib)
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, T_grid, Y, cmap='viridis')

ax.set_title('Diagrama de Fase: Onda Progresiva')
ax.set_xlabel('Posición (m)')
ax.set_ylabel('Tiempo (s)')
ax.set_zlabel('Desplazamiento (m)')

plt.show()

# Graficar la onda en función de la posición
plt.figure(figsize=(14, 6))

# Gráfica 1: Onda Progresiva en función de la posición
plt.subplot(1, 2, 1)
plt.plot(x, A * np.sin(k * x - omega * 0), label='t = 0 s')  # Tiempo inicial
plt.title('Onda Progresiva en el Espacio')
plt.xlabel('Posición (m)')
plt.ylabel('Desplazamiento (m)')
plt.grid(True)
plt.legend()

# Gráfica 2: Desplazamiento en función del tiempo para un punto fijo
x_fixed = longitud_onda / 4  # Posición fija para graficar
y_fixed = A * np.sin(k * x_fixed - omega * t)
plt.subplot(1, 2, 2)
plt.plot(t, y_fixed, label=f'x = {x_fixed:.2f} m')
plt.title('Onda Progresiva en el Tiempo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Desplazamiento (m)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

# Resultados
print(f"Tensión en el cordón: {ten:.2f} N")
print(f"Densidad lineal del cordón: {mu:.2f} kg/m")
print(f"Velocidad de la onda: {v:.2f} m/s")
print(f"Frecuencia de la onda: {f:.2f} Hz")
print(f"Número de onda: {k:.2f} rad/m")
print(f"Frecuencia angular: {omega:.2f} rad/s")

````
