# Introducción a HTML

## 1. Introducción y Definiciones
### ¿Qué es HTML?
HTML (HyperText Markup Language) es el lenguaje estándar para la creación de páginas web. Utiliza **etiquetas** para definir la estructura y el contenido de un documento web.

### Diferencia entre HTML, CSS y JavaScript
- **HTML**: Define la estructura y el contenido de una página web.
- **CSS**: Se utiliza para el diseño y la presentación visual.
- **JavaScript**: Añade interactividad y lógica al comportamiento de la página.

### Breve Historia de HTML
- Creado por Tim Berners-Lee en 1991.
- Evolucionó con diferentes versiones; actualmente usamos **HTML5**, que es más flexible y potente.

---

## 2. HTTP y Peticiones Web
### ¿Qué es HTTP?
HTTP (HyperText Transfer Protocol) es el protocolo que permite la comunicación entre navegadores y servidores web. Es el corazón de la web.

### Métodos Comunes de HTTP
- **GET**: Solicita datos desde un servidor (como una página HTML).
- **POST**: Envía datos al servidor (como un formulario).

### Cómo Funciona la Comunicación
1. El navegador envía una **petición HTTP** al servidor.
2. El servidor responde con el contenido solicitado (p. ej., una página HTML).
3. El navegador muestra el contenido al usuario.

---

## 3. Estructura de una Página HTML
La estructura básica de un documento HTML incluye:
- **Declaración DOCTYPE**: Especifica la versión de HTML.
- **Elemento `<html>`**: El contenedor raíz de la página.
- **Elemento `<head>`**: Contiene metadatos como el título.
- **Elemento `<body>`**: Contiene el contenido visible.

Ejemplo básico:

```html
<!DOCTYPE html>
<html>
    <head>
        <title>Mi Primera Web</title>
    </head>
    <body>
        <h1>¡Hola Mundo!</h1>
    </body>
</html>


## 4 Crear tu Primera Web: Paso a Paso

### 1. Preparativos
- Necesitas un editor de texto como **Visual Studio Code**, **Notepad++** o incluso el Bloc de notas.
- También necesitas un navegador web como **Chrome**, **Firefox** o **Edge**.

### 2. Escribir tu Página
1. Abre tu editor de texto.
2. Escribe el siguiente código básico de HTML:

```html
<!DOCTYPE html>
<html>
    <head>
        <title>Mi Primera Web</title>
    </head>
    <body>
        <h1>¡Hola Mundo!</h1>
    </body>
</html>


## 5 Ejemplo Básico sin Estilos

Aquí tienes un ejemplo más completo de una página HTML sencilla:

```html
<!DOCTYPE html>
<html>
    <head>
        <title>Mi Página Simple</title>
    </head>
    <body>
        <h1>Bienvenido a mi página</h1>
        <p>Este es un ejemplo de una página básica sin estilos.</p>
        <a href="https://www.google.com">Ir a Google</a>
    </body>
</html>    </body>
</html>



  ## Retos Simples
  
  ### Reto 1: Agregar un Encabezado Adicional
- Añade un encabezado `<h2>` con un mensaje personalizado.

```html
    <h2>Subtítulo de Bienvenida</h2>
    ```

    ### Reto 2: Crear una Lista
    - Incluye una lista ordenada o desordenada.

    ```html
    <ul>
        <li>Primera tarea</li>
        <li>Segunda tarea</li>
        <li>Tercera tarea</li>
    </ul>
    ```

    ### Reto 3: Crear un Enlace que Abra en una Nueva Pestaña
    - Usa el atributo `target ="_blank" ` para abrir un enlace en una nueva pestaña.

    ```html
    <a href="https://www.wikipedia.org" target="_blank">Ir a Wikipedia</a>
    ```
