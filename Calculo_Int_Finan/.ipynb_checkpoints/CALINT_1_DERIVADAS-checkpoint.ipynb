### 1. Límite de una Función

#### Definición
El límite de una función describe el comportamiento de dicha función a medida que sus argumentos se acercan a un valor específico. En otras palabras, nos indica hacia dónde se dirige el valor de la función cuando la variable independiente se aproxima a un número determinado.

Si $$f(x)$$ es una función, entonces:

$$\[
\lim_{{x \to c}} f(x) = L
\]$$

Esto significa que a medida que $$ (x)$$ se acerca al valor $$(c)$$, $$f(x)$$ se acerca al valor $$L$$.

#### Ejemplo 1: Límite sencillo

Consideremos la función $$ f(x) = 2x + 3 $$. Vamos a calcular el límite cuando $$(x)$$ tiende a 1.

$$
lim_{{x \to 1}} (2x + 3) = 2(1) + 3 = 5
$$

**Interpretación:** A medida que $$(x)$$ se acerca a 1, el valor de la función se acerca a 5.

#### Ejemplo 2: Límite con indeterminación

Consideremos ahora la función $$( f(x) = \frac{x^2 - 1}{x - 1} )$$. Calcularemos el límite cuando $$( x )$$ tiende a 1.

$$[
\lim_{{x \to 1}} \frac{x^2 - 1}{x - 1}
]$$

Este límite presenta una indeterminación \( \frac{0}{0} \). Sin embargo, al factorizar el numerador:

$$[
f(x) = \frac{(x - 1)(x + 1)}{x - 1}
]$$

Podemos simplificar:

$$[
f(x) = x + 1
]$$

Entonces:

$$[
\lim_{{x \to 1}} (x + 1) = 1 + 1 = 2
]$$

**Interpretación:** Aunque la función original presenta una indeterminación en $$( x = 1 )$$, su límite existe y es 2.

### 2. Derivada de una Función

#### Definición
La derivada de una función $$( f(x) )$$ en un punto $$( x )$$ mide la tasa de cambio instantánea de la función con respecto a su variable independiente en ese punto. Matemáticamente, la derivada de $$( f(x) )$$ respecto a $$( x )$$ se denota como $$( f'(x) )$$ o $$( \frac{d}{dx}f(x) )$$.

Para una función cuadrática general \( f(x) = ax^2 + bx + c \), la derivada se calcula como:

$$[
f'(x) = \frac{d}{dx}(ax^2 + bx + c) = 2ax + b
]$$

#### Ejemplo 1: Derivada de una función cuadrática simple

Consideremos la función $$( f(x) = 3x^2 + 4x + 5 )$$.

1. Aplicamos la regla de la potencia para derivar cada término:

$$[
f'(x) = \frac{d}{dx}(3x^2) + \frac{d}{dx}(4x) + \frac{d}{dx}(5)
]$$

2. Derivando cada término:

$$[
f'(x) = 6x + 4
]$$

**Interpretación:** La derivada $$( f'(x) = 6x + 4 )$$ nos indica la pendiente de la recta tangente a la curva en cualquier punto $$( x )$$.

#### Ejemplo 2: Derivada y pendiente en un punto específico

Consideremos ahora la función $$( f(x) = 2x^2 - 3x + 1 )$$. Vamos a encontrar la derivada y calcular la pendiente en $$( x = 2 )$$.

1. Derivamos la función:

$$[
f'(x) = 4x - 3
]$$

2. Evaluamos la derivada en $$( x = 2 )$$:

$$[
f'(2) = 4(2) - 3 = 8 - 3 = 5
]$$

**Interpretación:** La pendiente de la curva en $$( x = 2 )$$ es 5, lo que significa que la función está aumentando en ese punto.

#### Ejemplo 3: Derivada de una función cuadrática con un término negativo

Consideremos la función $$( f(x) = -x^2 + 2x + 3 )$$.

1. Derivamos la función:

$$[
f'(x) = \frac{d}{dx}(-x^2) + \frac{d}{dx}(2x) + \frac{d}{dx}(3)
]$$

2. Derivando cada término:

$$[
f'(x) = -2x + 2
]$$

**Interpretación:** La derivada $$( f'(x) = -2x + 2 )$$ indica que la pendiente de la curva disminuye a medida que $$( x )$$ aumenta, lo que refleja una parábola con abertura hacia abajo.

## Modelado de Función de Ingreso

A continuación, se presenta una tabla con datos de una empresa que muestra cómo el precio de venta de un producto disminuye a medida que se venden más unidades:

|----------------------|-------------------------|----------------------|
| 1                    | 48                      | ?                    |
| 2                    | 46                      | ?                    |
| 3                    | 44                      | ?                    |
| 4                    | 42                      | ?                    |

**Preguntas para los estudiantes:**

1. **Identifica la relación lineal:** Observa cómo el precio por unidad $$( p(x) )$$ disminuye a medida que se venden más productos. Escribe una ecuación que represente esta relación lineal.

2. **Deriva la función de ingreso $$( R(x) )$$:** Utiliza la relación entre $$( p(x) )$$ y $$( x )$$ para escribir la función de ingreso $$( R(x) = p(x) \times x )$$.

3. **Calcula el Ingreso Total:** Completa la columna de "Ingreso Total (R(x))" usando la función de ingreso que has derivado.

**Solución:**

1. **Relación lineal:** Observando los datos, el precio por unidad disminuye en 2 unidades monetarias por cada incremento en la cantidad vendida. Por lo tanto, la relación lineal es:

   $$[
   p(x) = 50 - 2x
   ]$$

2. **Función de ingreso:** Usando la relación lineal, la función de ingreso se puede expresar como:

   $$[
   R(x) = p(x) \times x 
   ]$$
   $$[= (50 - 2x) \times x 
   ]$$
   $$[= 50x - 2x^2
   ]$$

3. **Ingreso Total:** Usando la función de ingreso, calculamos los valores de la columna "Ingreso Total":

### 3. Aplicación de la Derivada en Costos

#### Introducción

En economía y administración, es común utilizar la derivada para optimizar funciones de costo y de ingreso. Un caso típico es encontrar el precio de venta que maximiza el ingreso o minimiza el costo. Para ello, se usa la derivada para identificar los puntos críticos (máximos y mínimos) de la función de costos.

#### Ejemplo: Maximización del Ingreso

Supongamos que una empresa tiene la siguiente función de ingreso $$( R(x) )$$, donde $$( x )$$ es la cantidad de productos vendidos, y el precio de venta de cada producto es $$( p(x) )$$:

$$[
R(x) = p(x) \times x
]$$

Supongamos que el precio $$( p(x) )$$ disminuye a medida que se venden más productos, siguiendo una relación lineal: $$( p(x) = 50 - 2x )$$. Entonces, la función de ingreso es:

$$[
R(x) = (50 - 2x) \times x = 50x - 2x^2
]$$

#### Paso 1: Derivada de la Función de Ingreso

Para maximizar el ingreso, necesitamos encontrar la derivada de $$( R(x) )$$ y luego determinar en qué valor de $$( x )$$ (cantidad de productos vendidos) se maximiza el ingreso.

$$[
R'(x) = \frac{d}{dx}(50x - 2x^2)
]$$

$$[
R'(x) = 50 - 4x
]$$

#### Paso 2: Encontrar el Punto Crítico

El punto crítico se encuentra igualando la derivada a cero y resolviendo para \( x \):

$$[
50 - 4x = 0
]$$

$$[
4x = 50
]$$

$$[
x = \frac{50}{4} = 12.5
]$$

Dado que la cantidad de productos $$( x )$$ debe ser un número entero, analizamos $$( x = 12 )$$ y $$( x = 13 )$$.

#### Paso 3: Evaluación de la Función de Ingreso

Evaluamos la función de ingreso para $$( x = 12)$$ y $$( x = 13 )$$:

$$[
R(12) = 50(12) - 2(12)^2 = 600 - 288 = 312
]$$

$$[
R(13) = 50(13) - 2(13)^2 = 650 - 338 = 312
]$$

Ambos valores dan el mismo ingreso, por lo que el ingreso máximo se obtiene al vender 12 o 13 unidades.

#### Interpretación

El precio de venta máximo que maximiza el ingreso se obtiene cuando se venden 12 o 13 productos. Al fijar el precio por unidad en $$( p(12) = 50 - 2(12) = 26 )$$ y $$( p(13) = 50 - 2(13) = 24 )$$, se maximiza el ingreso total a 312 unidades monetarias.

#### Conclusión

La derivada nos permite encontrar los puntos críticos de una función de costos o ingresos, lo que ayuda a tomar decisiones de precios estratégicas que maximicen las ganancias. Este procedimiento es fundamental en la optimización económica.

## Ejercicio 1: Maximización del Ingreso

### Enunciado:

Supongamos que una empresa tiene la siguiente tabla con datos de precios y cantidades vendidas. La relación entre el precio y la cantidad vendida sigue un patrón lineal. Con estos datos, determine la función de ingreso, calcule su derivada, y encuentre el ingreso máximo.

### Tabla de Datos

| Cantidad Vendida (x) | Precio por Unidad (p(x)) |
|----------------------|-------------------------|
| 0                    | 40                     |
| 1                    | 38                     |
| 2                    | 36                     |
| 3                    | 34                     |
| 4                    | 32                     |
| 5                    | 30                     |

### Guía para Resolver:

1. **Deducir la función precio**: Observa cómo el precio disminuye a medida que la cantidad vendida aumenta. Encuentra la ecuación lineal que describe esta relación. Aquí, la relación es lineal, por lo que podemos usar la forma \( p(x) = m \cdot x + b \). 

   Con los datos:
   - Cuando \( x = 0 \), \( p(x) = 40 \).
   - Cuando \( x = 5 \), \( p(x) = 30 \).
   
   La pendiente \( m \) se calcula como:
   $$[
   m = \frac{(30 - 40)}{(5 - 0)} = -2
   ]$$
   La función de precio es:
   $$[
   p(x) = -2x + 40
   ]$$

2. **Definir la función de ingreso**: Usa la relación encontrada para escribir la función de ingreso $$( R(x) = p(x) \cdot x )$$.
   
   $$[
   R(x) = (-2x + 40) \cdot x = -2x^2 + 40x
   ]$$

4. **Derivar la función de ingreso**: Encuentra la derivada de $$( R(x) )$$ con respecto a $$( x )$$.
   
   $$[
   R'(x) = \frac{d}{dx}(-2x^2 + 40x) = -4x + 40
   ]$$

5. **Encontrar el ingreso máximo**: Resuelve la derivada igualada a cero para encontrar el valor de $$( x )$$ que maximiza el ingreso.
   
   $$[
   -4x + 40 = 0 \implies x = 10
   ]$$
   
   Sustituye $$( x = 10 )$$ en la función de ingreso para encontrar el ingreso máximo:
   
   $$[
   R(10) = -2(10)^2 + 40 \cdot 10 = -200 + 400 = 200
   ]$$

6. **Resultado**:
   - **Cantidad óptima de productos a vender**: 10
   - **Ingreso máximo**: $200
