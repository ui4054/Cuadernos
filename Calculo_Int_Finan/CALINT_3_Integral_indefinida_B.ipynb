# 1. Método de Integración por Sustitución Simple

El **método de sustitución simple** se utiliza para simplificar una integral compleja transformando el integrando en términos de una nueva variable.

## Pasos para la Sustitución:
1. Identifica una expresión en el integrando que se pueda sustituir con una nueva variable $u$.
2. Encuentra la derivada de $u$ respecto de $x$, para obtener $du$ en términos de $dx$.
3. Reescribe la integral en términos de $u$ y $du$.
4. Integra en términos de $u$.
5. Deshaz la sustitución para devolver la solución a la variable original $x$.

## Ejemplo 1: 
Resolver la siguiente integral:
$$
\int (2x + 3) e^{x^2 + 3x} \, dx
$$

### Solución:

1. **Paso 1**: Definimos $u = x^2 + 3x$. Luego, derivamos para obtener:
   $$
   \frac{du}{dx} = 2x + 3 \implies du = (2x + 3) dx
   $$

2. **Paso 2**: La integral se transforma en:
   $$
   \int e^u \, du
   $$

3. **Paso 3**: Integramos en términos de $u$:
   $$
   \int e^u \, du = e^u + C
   $$

4. **Paso 4**: Volvemos a la variable original:
   $$
   e^{x^2 + 3x} + C
   $$

---

## Ejemplo 2: 
Resolver la siguiente integral:
$$
\int x \cos(x^2) \, dx
$$

### Solución:

1. **Paso 1**: Definimos $u = x^2$, de donde:
   $$
   \frac{du}{dx} = 2x \implies du = 2x \, dx
   $$
   Entonces, $\frac{du}{2} = x \, dx$.

2. **Paso 2**: La integral se convierte en:
   $$
   \frac{1}{2} \int \cos(u) \, du
   $$

3. **Paso 3**: Integramos:
   $$
   \frac{1}{2} \sin(u) + C
   $$

4. **Paso 4**: Volvemos a la variable original:
   $$
   \frac{1}{2} \sin(x^2) + C
   $$

---

# 2. Integración por Partes

La **integración por partes** es un método de integración basado en la regla del producto para la diferenciación. Se utiliza para integrar el producto de dos funciones. 

## Fórmula de Integración por Partes

La fórmula se expresa como:

$$
\int u \, dv = uv - \int v \, du
$$

Donde:
- \( u \) es una función que se elige para derivar.
- \( dv \) es la parte restante de la integral que se elige para integrar.

### Ejemplo 1

**Integrar** $ \int x \, e^x \, dx $.

**Paso 1:** Elegir $u $ y $ dv $:
- $ u = x $  $ \Rightarrow du = dx $
- $ dv = e^x \, dx $  $ \Rightarrow v = e^x $

**Paso 2:** Aplicar la fórmula:
$$
\int x \, e^x \, dx = x \, e^x - \int e^x \, dx
$$
$$
= x \, e^x - e^x + C
$$

**Resultado:**
$$
\int x \, e^x \, dx = e^x (x - 1) + C
$$

### Ejemplo 2

**Integrar** $\int \ln(x) \, dx$.

**Paso 1:** Elegir $ u $ y $ dv $:
- $u = \ln(x)$  $\Rightarrow du = \frac{1}{x} \, dx$
- $dv = dx$  $\Rightarrow v = x$

**Paso 2:** Aplicar la fórmula:
$$
\int \ln(x) \, dx = x \ln(x) - \int x \cdot \frac{1}{x} \, dx
$$
$$
= x \ln(x) - \int 1 \, dx
$$
$$
= x \ln(x) - x + C
$$

**Resultado:**
$$
\int \ln(x) \, dx = x \ln(x) - x + C
$$
