# 1. Método de Integración por Fracciones Parciales

El **método de fracciones parciales** es una técnica para descomponer una fracción racional en una suma de fracciones más simples, lo que facilita su integración. Este método es útil cuando se tiene una función racional, es decir, una fracción del tipo:

$$
\frac{P(x)}{Q(x)}
$$

Donde $ P(x)$ y $ Q(x)$ son polinomios, y el grado de $ P(x) $ es menor que el grado de $ Q(x)$.

### Pasos para la descomposición en fracciones parciales:

1. **Factorizar el denominador** $ Q(x) $ en factores lineales o cuadráticos irreducibles.
   
   Ejemplo: Supongamos que $ Q(x) = (x+1)(x-2) $, entonces la fracción se descompondrá en sumas del tipo:
   
   $$
   \frac{P(x)}{(x+1)(x-2)} = \frac{A}{x+1} + \frac{B}{x-2}
   $$

2. **Plantear la descomposición de la fracción** como una suma de fracciones parciales. Cada factor lineal en \( Q(x) \) se convierte en una fracción con una constante desconocida en el numerador. Para factores cuadráticos irreducibles, el numerador es una función lineal.

3. **Resolver el sistema de ecuaciones**: Multiplica ambos lados de la ecuación por el denominador original $ Q(x) $ para eliminar los denominadores. Luego iguala los coeficientes de los términos de igual potencia para encontrar los valores de $ A $, $ B $, etc.

4. **Integrar cada fracción parcial** por separado: Una vez que la fracción ha sido descompuesta, puedes integrar cada una de las fracciones parciales.

### Ejemplo:

Encuentra la integral:

$$
\int \frac{2x+3}{(x-1)(x+2)} dx
$$

1. **Factorización**: El denominador ya está factorizado como \( (x-1)(x+2) \).

2. **Descomposición**:
   
   $$
   \frac{2x+3}{(x-1)(x+2)} = \frac{A}{x-1} + \frac{B}{x+2}
   $$

3. **Resolver para $ A $ y $ B $**:
   Multiplicamos ambos lados por $(x-1)(x+2)$:
   
   $$
   2x + 3 = A(x+2) + B(x-1)
   $$
   
   Expandimos:
   
   $$
   2x + 3 = A(x) + 2A + B(x) - B
   $$
   
   Agrupamos términos:
   
   $$
   2x + 3 = (A+B)x + (2A - B)
   $$
   
   Ahora, igualamos los coeficientes de los términos de igual potencia:
   
   Para $ x $: $ A + B = 2 $ 
   Para el término independiente: $( 2A - B = 3 $

   Resolviendo este sistema, obtenemos $ A = 1 $ y $ B = 1 $.

4. **Reescribir la fracción**:

   $$
   \frac{2x+3}{(x-1)(x+2)} = \frac{1}{x-1} + \frac{1}{x+2}
   $$

5. **Integrar**:

   $$
   \int \frac{2x+3}{(x-1)(x+2)} dx = \int \frac{1}{x-1} dx + \int \frac{1}{x+2} dx
   $$

   El resultado  es:

   $$
   \ln|x-1| + \ln|x+2| + C = \ln| \frac{x-1}{x+2} | + C
   $$



## 2. Ejemplo de Integración por Fracciones Parciales

$$
\int \frac{3x^2 + 5x + 2}{(x-1)(x^2+4)} dx
$$

### Pasos:

1. **Factorización del Denominador**: El denominador ya está factorizado como $(x-1)(x^2 + 4)$. Tenemos un factor lineal $(x-1)$ y un factor cuadrático irreducible $(x^2 + 4)$.

2. **Descomposición en fracciones parciales**: La fracción se descompone en:

$$
\frac{3x^2 + 5x + 2}{(x-1)(x^2 + 4)} = \frac{A}{x-1} + \frac{Bx + C}{x^2 + 4}
$$

Donde $A$, $B$, y $C$ son constantes a determinar.

3. **Multiplicación por el denominador común**: Multiplicamos ambos lados de la ecuación por $(x-1)(x^2 + 4)$:

$$
3x^2 + 5x + 2 = A(x^2 + 4) + (Bx + C)(x-1)
$$

4. **Expansión**:

Primero expandimos el lado derecho:

$$
A(x^2 + 4) = Ax^2 + 4A
$$

Y

$$
(Bx + C)(x-1) = Bx^2 - Bx + Cx - C
$$

Entonces:

$$
3x^2 + 5x + 2 = Ax^2 + 4A + Bx^2 - Bx + Cx - C
$$

5. **Agrupamos términos**:

$$
3x^2 + 5x + 2 = (A + B)x^2 + (-B + C)x + (4A - C)
$$

6. **Igualar coeficientes**: Ahora, igualamos los coeficientes de los términos de igual potencia en $x$:

- Para $x^2$: $A + B = 3$
- Para $x$: $-B + C = 5$
- Para el término constante: $4A - C = 2$

7. **Resolver el sistema de ecuaciones**:

Primero, resolvemos la ecuación $A + B = 3$:

$$
B = 3 - A
$$

Sustituimos esto en la segunda ecuación $-B + C = 5$:

$$
-(3 - A) + C = 5 \implies -3 + A + C = 5 \implies A + C = 8
$$

Y en la tercera ecuación $4A - C = 2$:

$$
4A - C = 2
$$

Sustituyendo $C = 8 - A$:

$$
4A - (8 - A) = 2 \implies 4A - 8 + A = 2 \implies 5A = 10 \implies A = 2
$$

Ahora que tenemos $A = 2$, calculamos $B$ y $C$:

$$
B = 3 - 2 = 1
$$
$$
C = 8 - 2 = 6
$$

8. **Reescribir la fracción**: Ahora que tenemos los valores de $A$, $B$, y $C$, reescribimos la fracción como:

$$
\frac{3x^2 + 5x + 2}{(x-1)(x^2 + 4)} = \frac{2}{x-1} + \frac{x + 6}{x^2 + 4}
$$

9. **Integración**: Ahora integramos cada término por separado:

$$
\int \frac{2}{x-1} dx + \int \frac{x+6}{x^2+4} dx
$$

- La primera integral es simple:

$$
\int \frac{2}{x-1} dx = 2 \ln|x-1|
$$

- Para la segunda, dividimos la integral en dos partes:

$$
\int \frac{x}{x^2+4} dx + \int \frac{6}{x^2+4} dx
$$

La primera integral la resolvemos usando la sustitución $u = x^2 + 4$, con $du = 2x dx$, y obtenemos:

$$
\frac{1}{2} \ln|x^2 + 4|
$$

La segunda integral es una forma estándar, cuya solución es:

$$
\int \frac{6}{x^2+4} dx = 3 \arctan\left(\frac{x}{2}\right)
$$

10. **Solución final**: Sumamos todas las partes y obtenemos:

$$
2 \ln|x-1| + \frac{1}{2} \ln|x^2 + 4| + 3 \arctan\left(\frac{x}{2}\right) + C
$$
