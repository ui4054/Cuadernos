# Integrales por Sustitución Trigonométrica

## Introducción
Las integrales por sustitución trigonométrica son una técnica útil para resolver integrales que involucran expresiones cuadráticas. En este cuaderno, exploraremos los conceptos básicos y algunos ejemplos prácticos.

## Conceptos Básicos
Cuando tenemos integrales de la forma:

$$
\int \sqrt{a^2 - x^2} \, dx, \quad \int \sqrt{a^2 + x^2} \, dx, \quad \int \sqrt{x^2 - a^2} \, dx
$$

podemos utilizar las siguientes sustituciones trigonométricas:

1. **Para** $\sqrt{a^2 - x^2}$:
   - Sustitución: $x = a \sin(\theta)$
   - Diferencial: $dx = a \cos(\theta) d\theta$

2. **Para** $\sqrt{a^2 + x^2}$:
   - Sustitución: $x = a \tan(\theta)$
   - Diferencial: $dx = a \sec^2(\theta) d\theta$

3. **Para** $\sqrt{x^2 - a^2}$:
   - Sustitución: $x = a \sec(\theta)$
   - Diferencial: $dx = a \sec(\theta) \tan(\theta) d\theta$



## Ejemplo 1: Integral de la forma $\sqrt{a^2 - x^2}$

### Integral
Resolvamos la integral:

$$
\int \sqrt{16 - x^2} \, dx
$$

### Paso 1: Sustitución
Usamos la sustitución:

$$
x = 4 \sin(\theta) \quad \Rightarrow \quad dx = 4 \cos(\theta) d\theta
$$

### Paso 2: Sustituyendo en la Integral
Sustituyendo en la integral, obtenemos:

$$
\int \sqrt{16 - (4 \sin(\theta))^2} \cdot 4 \cos(\theta) d\theta = \int \sqrt{16(1 - \sin^2(\theta))} \cdot 4 \cos(\theta) d\theta
$$

### Paso 3: Simplificación
Esto se simplifica a:

$$
\int \sqrt{16 \cos^2(\theta)} \cdot 4 \cos(\theta) d\theta = \int 4 \cdot 4 \cos^2(\theta) d\theta = 16 \int \cos^2(\theta) d\theta
$$

### Paso 4: Resolviendo la Integral
Usamos la identidad:

$$
\cos^2(\theta) = \frac{1 + \cos(2\theta)}{2}
$$

Sustituyendo, tenemos:

$$
16 \int \cos^2(\theta) d\theta = 16 \int \frac{1 + \cos(2\theta)}{2} d\theta = 8 \int (1 + \cos(2\theta)) d\theta
$$

Resolviendo la integral:

$$
8 \left( \theta + \frac{\sin(2\theta)}{2} \right) + C = 8\theta + 4\sin(2\theta) + C
$$

### Paso 5: Regresando a la variable original
Sustituimos de vuelta usando $x = 4 \sin(\theta)$:

$$
\theta = \arcsin\left(\frac{x}{4}\right)
$$

Entonces, $\sin(2\theta) = 2\sin(\theta)\cos(\theta) = 2 \cdot \frac{x}{4} \cdot \sqrt{1 - \left(\frac{x}{4}\right)^2} = \frac{x}{2} \cdot \sqrt{1 - \frac{x^2}{16}}$.

Finalmente, la solución es:

$$
8 \arcsin\left(\frac{x}{4}\right) + 2x \sqrt{4 - x^2} + C
$$


## Ejemplo 2: Integral de la forma $\sqrt{a^2 + x^2}$

### Integral
Resolvamos la integral:

$$
\int \sqrt{4 + x^2} \, dx
$$

### Paso 1: Sustitución
Usamos la sustitución:

$$
x = 2 \tan(\theta) \quad \Rightarrow \quad dx = 2 \sec^2(\theta) d\theta
$$

### Paso 2: Sustituyendo en la Integral
Sustituyendo en la integral, obtenemos:

$$
\int \sqrt{4 + (2 \tan(\theta))^2} \cdot 2 \sec^2(\theta) d\theta = \int \sqrt{4(1 + \tan^2(\theta))} \cdot 2 \sec^2(\theta) d\theta
$$

### Paso 3: Simplificación
Esto se simplifica a:

$$
\int \sqrt{4 \sec^2(\theta)} \cdot 2 \sec^2(\theta) d\theta = \int 4 \sec^3(\theta) d\theta
$$

### Paso 4: Resolviendo la Integral
La integral $\int \sec^3(\theta) d\theta$ se puede resolver usando la fórmula:

$$
\int \sec^3(\theta) d\theta = \frac{1}{2}(\sec(\theta) \tan(\theta) + \ln | \sec(\theta) + \tan(\theta) |) + C
$$

Por lo tanto, tenemos:

$$
4 \cdot \frac{1}{2}(\sec(\theta) \tan(\theta) + \ln | \sec(\theta) + \tan(\theta) |) + C = 2(\sec(\theta) \tan(\theta) + \ln | \sec(\theta) + \tan(\theta) |) + C
$$

### Paso 5: Regresando a la variable original
Sustituimos de vuelta usando $x = 2 \tan(\theta)$:

$$
\sec(\theta) = \sqrt{1 + \tan^2(\theta)} = \frac{\sqrt{4 + x^2}}{2}
$$

$$
\tan(\theta) = \frac{x}{2}
$$

Finalmente, la solución es:

$$
\sqrt{4 + x^2} \cdot \frac{x}{2} + 2 \ln \left| \sqrt{4 + x^2} + x \right| + C
$$


## Ejemplo 3: Integral de la forma $\sqrt{x^2 - a^2}$

### Integral
Resolvamos la integral:

$$
\int \sqrt{x^2 - 9} \, dx
$$

### Paso 1: Sustitución
Usamos la sustitución:

$$
x = 3 \sec(\theta) \quad \Rightarrow \quad dx = 3 \sec(\theta) \tan(\theta) d\theta
$$

### Paso 2: Sustituyendo en la Integral
Sustituyendo en la integral, obtenemos:

$$
\int \sqrt{(3 \sec(\theta))^2 - 9} \cdot 3 \sec(\theta) \tan(\theta) d\theta = \int \sqrt{9(\sec^2(\theta) - 1)} \cdot 3 \sec(\theta) \tan(\theta) d\theta
$$

### Paso 3: Simplificación
Esto se simplifica a:

$$
\int 3 \cdot 3 \tan(\theta) \cdot 3 \sec(\theta) \tan(\theta) d\theta = 9 \int \tan^2(\theta) \sec(\theta) d\theta
$$

### Paso 4: Resolviendo la Integral
Usamos la identidad:

$$
\tan^2(\theta) = \sec^2(\theta) - 1
$$

Entonces, la integral se convierte en:

$$
9 \int \tan^2(\theta) \sec(\theta) d\theta = 9 \int (\sec^2(\theta) - 1) \sec(\theta) d\theta = 9 \left( \int \sec^3(\theta) d\theta - \int \sec(\theta) d\theta \right)
$$

Ambas integrales se pueden resolver usando las fórmulas ya mencionadas.

### Paso 5: Regresando a la variable original
Finalmente, sustituimos de vuelta usando $x = 3 \sec(\theta)$:

$$
\sec(\theta) = \frac{x}{3}
$$

La solución completa, después de sustituir y simplificar, será:

$$
\frac{x}{3} \sqrt{x^2 - 9} + 3 \ln \left| x + \sqrt{x^2 - 9} \right| + C
$$
